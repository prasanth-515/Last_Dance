```python
# 1. Arithmetic Operators
# Q1. Basic Arithmetic:  
# Take two numbers from the user and perform addition, subtraction, multiplication, division, and modulus.

a = eval(input('Enter a :'))
print("Enter a : ", a)
b = eval(input('Enter b :'))
print('Enter b : ', b)
c = input('Enter the Operator : ')
print(f'Enter the Operator (" {c} ")')
x = a+b
d = a-b
e = a*b
f = a/b
g = a//b
h = a%b
i = a**b

if c=='+':
    print(f'a + b = {x}')
elif c=='-':
    print(f'a - b = {d}')
elif c=='*':
    print(f'a * b = {e}')
elif c=='/':
    print(f'a / b = {f}')
elif c=='//':
    print(f'a // b = {g}')
elif c=='%':
    print(f'a % b = {h}')
elif c=='**':
    print(f'a ** b = {i}')
else:
    print('❌ Enter Valid Operator ❌')

```

```python
# Q2. Compare Two Numbers:  
# Check if one number is greater, smaller, or equal to another.

a = int(input("Enter a :"))
print('Enter value a : ', a)
b = int(input('Enter b :'))
print('Enter value b : ', b)

if a > b:
    print(f'a ({a}) is greater')
elif b > a:
    print(f'b ({b}) is greater')
else:
    print(f'a({a}) and b({b}) is equal to one another')
```

```python
# Q4. Compound Assignment:  
# Take a number and update it using +=, -=, *=, /=.

x =int(input('Enter a : '))
print('Enter the value of x : ',x)
y =int(input('Enter y : '))
print('Enter the value of y : ',y)
print()

x+=y
print(f'x+=y --> x is {x} and y is {y}')
print()

x-=y
print(f'x-=y --> x is {x} and y is {y}')
print()

x*=y
print(f'x*=y --> x is {x} and y is {y}')

x/=y
print(f'x/=y --> x is {x} and y is {y}')
```

```python
# Q5. Identity Check:  
# Check if two variables point to the same object using is and is not.

a = [23,23,23]
b = [23,23,23]

if a is b:
    print('a is b')
else:
    print('a is not b')
print()

if a == b:
    print('a == b')
else:
    print('a != b')
```

```python
# Q6. Membership Test:  
# Check if a character exists in a string.

# Enter string: Hello  
# Enter char: e  
# 'e' in "Hello" → True

a = str(input('Enter a string : '))
print('Enter a string : ',a)

b = str(input('Enter a char : '))
print('Enter a char : ',b)
print()

if b in a:
    print(f'Char ({b}) is in the given String ({a})')
elif b not in a:
    print(f'Char ({b}) is not in the given Stirng ({b})')
else:
    print('Enter a Valid string')
```

```python
# Q7 : Palindrome String:  
# Write a program to check if a given string is a palindrome (reads the same forward and backward).

# Example Run:

# Code
# Enter string: madam
# Output: Palindrome

# Enter string: python
# Output: Not Palindrome

a = str(input('Enter a string : '))
print('Enter a string : ', a )
print()

if a == a [::-1]:
    print(f'{a} is Palindrome')
else:
    print(f'{a} is not a Palindrome')
    
```

```python
# Q8. Case Conversion:  
# Take a string and print it in uppercase, lowercase, and title case.
# Output Example:

# Code
# Enter string: hello world  
# Upper → HELLO WORLD  
# Lower → hello world  
# Title → Hello World

a = str(input('Enter a string : '))
print("Enter a string : ", a)
print()
print(f'Upper -> {a.upper()}')
print("Lower -> {}".format(a.lower()))
print('Title -> ', a.title () )
```
