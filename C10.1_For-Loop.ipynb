### LOOP

```python
# A loop is used to execute a piece of code repeated.
# Whenever we want to execute a piece of code n-times, then we use loop.

# Types of Loops
# 1. for loop
# 2. while loop
```

```python
for a in range(0,11,2):
    print(f'Countdown({a}) : {a} ')
```

```python
# 1. for loop
# syntax - for positioner in iterable:
#              code
#              code
# syntax - for loop iterates (traverses) upon the elements of the iterable element by element from the beginning to the end.
#        - the positioner is a variable that points to each element one by one.
```

```python
st = 'Prasanth'
print(st,type(st))
print()

for a in st:
    print(a,end='_')
print()
for a in range(len(st)):
    print('^ ',end='')
print()
for a in range(len(st)):
    print('| ',end='')
print()

# to print the indexes

for a in range(len(st)):
    print(a,end='-')
print()
print()

for a in range(len(st)):
    print(st[a],f'({a})')
print()

# to get it in reverse

for a in range(-1,-(len(st)+1),-1):
    print(st[a], f'({a})')
```

```python
# Q1. Print numbers 1 to 10

x = 11
for a in range(x):
    print(a, end='_')
print()

for a in range(x,0,-1):
    print(a, end='_')
print()

# Q2. Print even numbers between 1 to 20

x = 21
for a in range(1,x):
    if a%2==0:
        print(a,end='_')

print()

# Q8. Print only odd numbers between 1 to 50

x = 51
for a in range(1,x):
    if a%2!=0:
        print(a,end='_')


print()

# Q3. Print multiplication table of any number

x = int(input('Enter x :'))
print('Enter X(multiplicand) : ', x)
y = int(input('Enter y :'))
print('Enter y(multipler) : ', y)

for a in range(y+1):
    Q = x*a
    print(Q,end='_')
print()
    
```

```python
# Q4. Print each character of your name

a=str(input('Enter your word : '))
print('Enter you word : ',a)

for x in range (len(a)):
    print(a[x])
print()

for x in range (-1,-(len(a)+1),-1): #in reverse
    print(a[x])
```

```python
# Q10. Check if a number is Prime

x = int(input('Enter x :'))
print('Enter X : ', x)

for a in range(2,x-1):
    if x%a!=0:
        print(f'X({x}) is prime number')
    else:
        print(f'X({x}) is not prime number')
```

```python
import random

a = random.randint(1,20)
print('I have choosen a number between 1 to 20, Try guessing it')
print()

for x in range(1,11):  #allowing 10 attempts to guess
    y= int(input('Enter your guess : '))
    print(f'Attempt({x}). Your Guess is : {y}')
    if y==a:
        print('🎉😍You Got It😍🎉')
        break
    elif y>a:
        print('Higher, Keep Trying')
    else:
        print('smaller, Keep Trying')
else:
    print('Sorry you run out of attempts, The correct number is {}'.format(a))
        
```
