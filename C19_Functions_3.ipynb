```python
# reduce()

# - It is used to perform aggregation operations on the elements of an iterable.
# - Reduce() take two arguments and returns an aggregated value.
# - The first argument is a function and the second argument is an iterable.
# - The function that we pass inside the reduce() must have two arguments always.
# - To use reduce() we have to import it from the functools module.
```

```python
from functools import reduce

lst=[1,3,5,6,5,4,3]

def add(x,y):
    return x+y
s = reduce(add,lst)
print(s)
print()

a = reduce(lambda x,y:x+y, lst)
print(a)
```

```python
lo = [2,4,1,56,75,12,34,98,1,4,4]

def max(x,y):
    return x if x > y else y
max(34,65)

maxx = reduce(max,lo)
print(maxx)

def min(x,y):
    return x if x<y else y

minn = reduce(min,lo)
print(minn)
```

```python
lst=[1,3,5,6,5,4,3]

ab = reduce (lambda x,y : x + y, lst)
print(ab)
```

```python
# Properties of Functions.
# 1. Function Aliasing.
# 2. A fucnction can be passed as an argument to another function.
# 3. A function may be called inside another function.
# 4. A fucntion may be defined inside another function (Nested function)
# 5. A function may return another function.
```

```python
# 1. Function Aliasing.

# - When we assign a function to a variable (name) then we call it as function aliasing.
# - The variable acts as the function itself after function aliasing.

def tes():
    return 'Hello Python'
print(tes(), type(tes))   # we assigning a name to function

x = tes()   # assign a function to a variable.
print(x)
```

```python
# 2. A fucnction can be passed as an argument to another function.

def add (x,y):
    z = x + y
    # print(z)
    return z
x = add(5,4)

def te (a):
    print(a,type(a))
te(6)
print()
te(add(23,23))
te(x)
```

```python
# Create a function perform_operation that takes 3 arguments.
# The first argument is a function and the rest two arguments are values.
# The perform_operation() should perform the operation according to the function passed and should return the result.

def add(x,y):
    return x+y
def sub(x,y):
    return x-y
def mul(x,y):
    return x*y
def div(x,y):
    return x/y


def oper(a,x,y):
    return(a(x,y))

oper(add,34,32)
oper(sub,23,12)
```

```python
# 3. A function may be called inside another function.

def sq(x):
    return x ** 2

def add(x,y):
    return sq(x+y)

add(2,3)
```

```python
# 4. A fucntion may be defined inside another function (Nested function)

def outer():
    def inner():
        pass

def f1():
    print('I am the outer function')
    def f2():
        print('I am the inner function')
    f2()
f1()  # f1 is global and f2 is local
```

```python
# 5. A function may return another function.

def ff1():
    print('I am ff1 function')
    def ff2():
        return 'I am ff2 function'
    return ff2()

# ff1()
x = ff1()
print(x)
```

```python
# Recursive function
# - A function that calls itself is a recursive function.
# - when the function call of a function is written within its own definition then such a function is known as a recursive function.

def rev():
    rev()

def pa(x):
    password = '1234pa'
    if x == password:
        print('Password Matched')
        print('Log in Successful')
    else:
        print('Password is incorrect')
        x = input('Re-enter Password : ')
        print('Re-enter a password : ', x)
        pa(x) # function that calls itself.

x = input('Enter a Password : ')
print('Enter a Password : ',x)
res = pa(x)
```

```python
# create a function fact that takes a number as argument and return its factorial.

def fac():
    a = int(input('Enter a : '))
    print('Enter a number : ',a)
    x = 1
    for i in range(1,a+1):
        x = x * i
    return f'The factorial of {a} is : {x}'

o = fac()
print(o)
```

```python
def pr_lst(a):
    res = []
    for i in range(2,a+1):
        for j in range(2,i):
            if i % j == 0:
                break
        else:
            res.append(i)
    return res

a = int(input('Enter a : '))
print('Enter a : ',a)
yy = pr_lst(a)
print(yy)

```
