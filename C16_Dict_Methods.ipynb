```python
# 1. keys()
# - It returns a dict_keys object that contains a list of keys.

emp = {'Name':'Prasanth','Age':30,'City':'Salem','Ph':12345}
print(emp.keys(),type(emp))

emp_details = {'emp_1':{'Name':'Prasanth','Age':30,'City':'Salem','Ph':123456},
               'emp_2' :{'Name':'Ram','Age':28,'City':'Coimbatore','Ph':876543}, 
               'emp_3' :{'Name':'Sam','Age':29,'City':'Chennai','Ph':8765431}}

print(emp_details.keys())
print()

for i in emp_details.keys():
    print(i,emp_details[i])
```

```python
# 2. values()
# - It returns a dict_values objct that contains a list of values.

print(emp.values())
print()
print(emp_details.values())
print()

for i in emp_details.values():
    print(i)
    for j in i:
        print(j)
```

```python
# 3. items()
# - It returns a dict_items object that contains a list of tuples.
# - Each tuple contains a 2 elemets, where the first elements is the key and the 2nd element is the value.

print(emp.items())
print()
print(emp_details.items())
print()

for i,j in emp_details.items():  # unpacking
    print(i)
    print(j)
    print(i,j)
print()

for i,j in emp.items():  # unpacking
    print(i,j)
```

```python
# 4. update()
# - It takes a dictionary as argument and updates its key:value pairs to the existing dictionary.

d1 = {1:100,2:200}
d2 = {3:300,4:400}

# d1.update(d2)
# - The items of d2 will be updated to d1.

d1.update(d2)
print(d1)
print()

d1 = {1:100,2:200}
d2 = {2:300,4:400}  # when it having common key.
d1.update(d2)
print(d1)
print()

# create a dictionary that has the numbers from 1 to 10 as keys and their squares as values.

d ={i:i**2 for i in range(1,11)}
print(d)
d.update({i:i**3 for i in range(1,11)})
print(d)
```

```python
# 5. get()
# - It takes a key as argument and returns the value associated to that key.
# - If the key is not available then it returns None.

emp = {'Name':'Prasanth','Age':30,'City':'Salem','Ph':12345,'Salary':50000}
print(emp.get('Name'))
print()
print(emp.get('Country')) # since key is not available it returns None.
print()

# - We can also provide a default value to return if the key is not available.
print(emp.get('Name', 'Not Available')) # returns 'Not Available' if key is not available.
print()
```

```python
# 6. setdefault()
# - IT takes a key as argument and returns the value at that particular key.
# - If the key is not available then it creates the key and maps to None (defalut value). 
# - For any invalid key if we pass value as well then it maps the key with the given value.

emp = {'Name':'Prasanth','Age':30,'City':'Salem','Ph':12345,'Salary':50000}

print(emp['Age']) # returns 30
print(emp.get('Age')) # returns 30
print(emp.setdefault('Age')) # returns 30
print()

print(emp.setdefault('Country')) # creates the key 'Country' with default value None
print(emp) # displays the updated dictionary
print()

print(emp.setdefault('Nationality','India')) # creates the key 'Nationality' with value 'India'
print(emp) # displays the updated dictionary
print()
```

```python
# 7. pop()
# - It takes a key as an argument and it returns the value at theat particular key and removes the key:value pair from the dictionary.
# - For any invalid key it raises KeyError.

emp = {'Name':'Prasanth','Age':30,'City':'Salem','Ph':12345,'Salary':50000}
print(emp)
print()

print(emp.pop('Age')) # returns 30 and removes the key:value pair from the dictionary
print(emp)
print()

# print(emp.pop('Country')) # raises KeyError since the key is not available
```

```python
# 8. popitem()
# - It returns the last key:value pair in a tuple and removes it from the dictionary.

emp = {'Name':'Prasanth','Age':30,'City':'Salem','Ph':12345,'Salary':50000}
print(emp)
print()
print(emp.popitem()) # returns the last key:value pair in a tuple and removes it from the dictionary
print(emp)
print()
print(emp.popitem()) # returns the last key:value pair in a tuple and removes it from the dictionary
print(emp)
```

```python
# 9. formkeys()
# - It takes an iterable and a value as arguments.
# - The elements of the iterable become the key and all the keys are mapped with the given value.

names = ['Prasanth','Ram','Sam','Ravi','Siva']
emp = dict.fromkeys(names, 'Employee')  # every key is mapped with the same value.
print(emp)
print()

for i in names:
    print(i, 'Star')
```

```python
# 10. copy()
# - It copies all the elements from a dictionary and returns a new dictionary with the same elements.

d = {'Name':'Prasanth','Age':30,'City':'Salem','Ph':12345,'Salary':50000}
d2 = d

print(d,id(d))
print(d2,id(d2))

d2 = d.copy()  # creates a new dictionary with the same elements.
print(d,id(d))  
d['Age'] = 31
print(d,id(d))  # id of d is same since it is updated.
print(d2,id(d2))  # id of d2 is different since it is a copy.
```

```python
# 11. clear()
# - It removes all the elements from a dictionary.

d = {'Name':'Prasanth','Age':30,'City':'Salem','Ph':12345,'Salary':50000}
print(d)
print()

d.clear()  # removes all the elements from the dictionary
print(d)  # displays an empty dictionary
```
