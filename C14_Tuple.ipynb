```python
# Tuple

# - A tuple is a collection of elements kept inside the () .
# - Each element of a tuple is seperated by a comma.
# - A tuple can contain heterogenous elements.
# - A tuple can contain duplicate elements as well.

t = (1,1,2,5.4,True,None,'HI')
print(t,type(t))
```

```python
# Operations on a Tuple.

# - 1. Concatenation
# - 2. Repetation
# - 3. Length
# - 4. Indexing and Slicing
# - 5. Membership
```

```python
# 1. Concatenation.

# - Joining two or more tuples together is known as tuple concatenation.
# - + operator is used to concatenate two tuples together.
# - A tuple can be concatenated with a tuple only.
# - If we try to concatenate a tuple with some other type then it raises TypeError.

a = (1,2,3,45,68)
b = (20,30,58,93)
c = a+b
print(c)
print()

x = ('Hi','Hello','Sam')
y = ('van','Bun')
z = x + y
t = c + z
print(z)
print()
print(t)

m = 'Hello'
n = (1,3,45,'sam')
# l = m+n   which will gives error..   since it is different type
```

```python
# 2. Repetation

# - * operator is used to repeat a tuple.

a = (1,2,3,45,68)
print(a * 3)
```

```python
# Indexing and Slicing

# Indexing.
# - A tuple is an ordered collection of data as each element of the tuple gets stored at a particular index.
# - A tuple is a sequence type as it supports index.
# - A tuple supprots +ve and -ve index both, same as list and string.

t = (10,'Sam',3.5,True,None,3+5j,[2,3,'sam'],{2:'sam',4:'City'})
print(t,type(t))
print()
print(t[-len(t)])   # -len(t) = -8
print()

#  slicing
# t[start_index:stop_index:step_size]
# step_size is 1

# to get first four elements.

t = (10,'Sam',3.5,True,None,3+5j,[2,3,'sam'],{2:'sam',4:'City'})
print(t[0:4:1])
print(t[0:4])
print('Reversed : ', t[-1::-1])
print('Reversed : ', t[::-1])
print(t[0::2])
```

```python
# Membership

# 1. in
# 2. not in

t = (10,'Sam',3.5,True,None,3+5j,[2,3,'sam'],{2:'sam',4:'City'})

print(23 in t)
print('hello' not in t)
```

```python
# How to create a tuple

# Case 1:
# - manually keep the elements inside the ()

# Case 2:
# - by using tuple () - by typecasting

# tuple(iterable) -> It returns a tuple. (Typecasting)

a = "Prasanth"
t = tuple(a)
print(t,type(t))
print()

lst = [1,2,34,'Hi']
t = tuple(lst)
print(t,type(t))
print()

emp = {1:'Age', 2:'Name', 3:'City'}
print(emp,type(emp))
t = tuple(emp)
print(t)
print(t,emp.values())
print()

r = range(1,11)
print(r,type(r))
print()
t = tuple(r)
print(t,type(t))
print()

e = enumerate('Prasanth')
print(e,type(e))
print()
t = tuple(e)
print(t,type(t))
print()

z = zip('abcd',['Hi','Hello','Bye','Come'])
print(z,type(z))
print()
t = tuple(z)
print(t,type(t))
print()

a = (2,4,2,56,7,6,2,3,9,0,13)
f = filter(lambda x:x%2==0, a)
print(f,type(f))
print()
t = tuple(f)
print(t,type(t))
print()

a = (2,4,2,56,7,6,2,3,9,0,13)
m = map(lambda x:x*5,a)
print(m,type(m))
print()
t = tuple(m)
print(t,type(t))
```

```python
# Case 3
# - Creating a tuple of single element.
# - When we pass a single element inside the () then it is not a tuple but the element itself.
# - To create a tuple of single element, we have to put the comma after the element inside the ()

a = ('Hello')
print(a,type(a))
print()

b = ('Hello',)
print(b,type(b))
```

```python
# Case_4

# - we can create a tuple without () as well.

# Packing
# - Putting or assigning multiple values to a single variable.
# - Packing is supported by tuple in python.
# - When we assign multiple values to a single variable then python internally creates a tuple and packs the values in that.

t = 10,20,30,40,'Sam',None
print(t,type(t))
print()
```

```python
# Unpacking
# - Assigning mulitple values from a container to different variables is known as unpacking.

a,b,c,d,e = (10,20,40,30,78)
print(a,type(a))
print()

t = (10,20,40,30,78)
a,b,c,d,e = t
print(t,type(t))
```

```python
# Case - 5
# - By taking input from the user.

x = eval(input('Enter tuple values : '))
print(x,type(x))
print()
```
