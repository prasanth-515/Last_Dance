```python
# Dictionary.
# - It is a collection of key:value pairs kept inside the {}
# - Each key:value pair is known as an item or element of the dictionary.
# - Each element or item is seperated by comma in dictionary.
# - Keys cannot be duplicated, but value can be duplicted.
# - If we have duplicate keys then it updates the latest mapped value to the key.

emp = {'Name':'Ravi','Age':23,'Sex': 'Male','City':'Salem','Name':'Kumar','name':'Sam'} # if the keys are same, then the last key value 
print(emp)                                                                              # will  be called at first key
print()

emp1 = {1:100,2:100,3:200,4:200}
print(emp1)
print()
```

```python
# Operations on a dictionary.

# Concatenation, repetition, indexing and slicing are not possible.
# Length and membership only possible.

emp = {'Name':'Ravi','Age':23,'Sex': 'Male','City':'Salem','Name':'Kumar','name':'Sam'}
print(len(emp))
print()

print('Yes it is') if 'Name' in emp else print ('Noooo')
```

```python
# Accessing a value from a dictionary

# - we need to enter a valid key value to access it.

emp = {'Name':'Ravi','Age':23,'Sex': 'Male','City':'Salem','Name':'Kumar','name':'Sam'}

print(emp['Name'],emp['Age'])
```

```python
# Adding a key value.
# d[key] = value
# - If the key doesn't exist int the dict, It adds the key:value pair at the end of the dict
# - If the key already exist in the dict then it modifies the value at that particular key.

emp = {'Name':'Ravi','Age':23 ,'Sex': 'Male','City':'Salem','Name':'Kumar','name':'Sam'}

emp['Phone'] = 23457218
print(emp)
```

```python
# Creating a dictonary

# Case_1
# - Manual creation.

# Case_2
# - By using dict funtion - typecasting.
# dict(iterable)

# - the iterable that we pass inside the dict() must have one more iterable inside it.
# - The inner iterable must have only 2 elements.
# - The first element becomes the key and the second element becomes the values.

# Output - {'Name' : 'Sam'}

d = dict([('Name','Sam')])
print(d,type(d))
print()

d = dict((('Name','Prasanth'),('Age',30)))
print(d)
print()
```

```python
# By using enumerate() inside the dict ()

# enumerate()
# - It takes an iterable as argument.
# - It maps the elements of the iterable with a sequence of numbers and returns an enumerate object.
# - When we pass an enumerate object inside the dict(), we get a dict.
# - By default it starts mapping with 0.
# - If we spcify any number then it starts mapping with that.


names = ['Ravi','Kumar','Shiva','Praveen']
e = enumerate(names)
print(e)
print()
l = list(e)
print(l)
print()

e = enumerate(names,11)   # mapping number we can declare here, if needed
print(e)                  # enumerate iteration can exhausted after using once.
print()
d = dict(e)
print(d)
print()

lst = ['Process-K','Process-L','Process-M','Process-N']
for num,p in enumerate (lst,1):
    print(num,p)
```

```python
# By using zip () within the dict()

# - It takes 2 or more iterables as argument.
# - It maps the elements of the iterable one by one and returns a zip object.
# - While creating the dict using zip(), it must have only 2 iterables.
# - The elements of the first iterable become the key and the elements of the second iterable become the value.

z = zip('abcd',[10,20,30,40])
print(z,type(z))
print()
print(list(z))
print()

z = zip('abcd',[10,20,30,40]) # for dict we should have only 2 elements, since it's pairing with values by keys.
print(z,type(z))
print()
d = dict(z)
print(d)
print()

a = [101,102,103,104,105]
b = ('Ram','Sam','Kumar','Ravi','Shiva')

if len(a) == len(b) :
    d = dict(zip(a,b))
    print(d)
else:
    print('Keys and values length is not equal')
```

```python
# Case_3

# - By taking input from user.

a = eval(input('Enter a dictionary : ')) # manual creation
print(a,type(a))

# Using loop

keys = []
values = []

for i in range(4):
    k = eval(input('Enter the key : '))
    v = eval(input('Enter the value : '))
    keys.append(k)
    values.append(v)
d = dict(zip(keys,values))   
print(d) 
```

```python
# Case_4
# - By using dictionary comprehension
# - It is a one liner code to create a dict.
# - It is an easier way to create a dict using loop.

# {key:value for loop}
# let key should be 1 to 10 and the value should be square of key
d = {i:i**2 for i in range (1,11)}
print(d)
print()

# Pair ASCII with values
d = { chr(i):i for i in range (97,123) }
print(d)
print()

# form tuple
d = {}
t = (1,3,5,2,8,7,5,3,9,0,1,2,5,7,9,9,2,3,4,5,7,8,9,3,6,7,2,9)
for i in t:
    if i%2==0:
        d[i]=i**2
print(d)
print()

# even should be squared and odd should be cubed
d = {}
for i in range(1,11):
    if i%2==0:
        d[i]=i**2
    else:
        d[i]=i**3
print(d,type(d))
print()

# same in ternary operator

d={ i:i**2 if i%2==0 else i**3 for i in range(1,11)}
print(d,type(d))
```
