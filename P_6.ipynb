```python
# Q1. Print numbers from N down to 1 (countdown)
# Enter N: 5
# 5 4 3 2 1
# Go! 🚀

n = int(input('Enter N : '))
print('Enter N: ', n)

while n > 0:
    print(n, end=' ')
    n -= 1
print()
print('Go! 🚀')
```

```python
# Q2. Keep asking user to guess a number until correct
# Secret number = 7
# Enter guess: 3 → Wrong!
# Enter guess: 9 → Wrong!
# Enter guess: 7 → Correct! 🎉

import random
r = random.randint(1,10)

n = 1
while n > 0:
    a = int(input('Enter guess : '))
    if a != r:
        print(f'Enter guess : {a} -> Wrong!')
    else:
        print(f'Enter guess : {a} -> Correct! 🎉')
        break
    n = n+1
```

```python
# Q3. Find sum of digits of a number
# Enter number: 1234
# 1+2+3+4 = 10

n = int(input('Enter number : '))
print('Enter number : ',n)

a = 0
while n > 0:
    c = n % 10
    a = a + c
    n = n // 10
print(a)
    
```

```python
# Q4. Reverse a number
# Enter number: 12345
# Reversed = 54321

n = int(input('Enter number : '))
print('Enter number : ',n)

b = 0
while n > 0:
    a = n % 10
    b = b*10 + a
    n = n // 10
print(f'Reversed = {b}')
```

```python
# Q5. Count how many times a digit appears in a number
# Enter number: 122333
# Enter digit to find: 3
# 3 appears 3 times

n = int(input('Enter number : '))
print('Enter number : ',n)
f = int(input('Enter digit to find : '))
print('Enter digit to find : ',f)
print()

x = 0
while n > 0:
    a = n % 10
    if f == a:
        x += 1
    n = n // 10
print(f'{f} appears {x} times')

```
