```python
## Linus Torvalds
##### Git is developed by Linus Torvalds, the creator of Linux, and is widely used for version control in software development.
##### He invented Linux in 1991.
##### He developed Git and Linus(Most powerful OS)
##### 96.3% of servers run by Linux.

## Git 
##### Git is a version control system that allows developers to track changes in their code, collaborate with others, and manage different versions of their projects.
##### git config --global user.name "Your Name" - type in cmd
##### git config --global user.email "your.email@example.com" - type in cmd

## Steps to use Git
##### 1. Change - change the code and save it.
##### 2. Add (staging) - add the changed code to the staging area - add "intro.ipynb"
##### 3. Commit - commit the changes to the local repository - commit -m "Initial commit"
##### 4. Push - push the changes to the remote repository - push origin main

### Terminal Commands
##### git init - Initializes a new Git repository in the current directory.
##### git add <file> - Adds the specified file to the staging area. 
##### git commit -m "message" - Commits the staged changes to the local repository with a message describing the changes.
##### git push origin <branch> - Pushes the committed changes to the remote repository on the specified branch.
##### git pull origin <branch> - Pulls the latest changes from the remote repository on the specified branch to the local repository.
##### git status - Shows the current status of the working directory and staging area, including any changes that have been made.
##### git diff - Shows the differences between the working directory and the staging area, or between commits.
##### git log - Shows the commit history of the repository, including commit messages, authors, and timestamps.

### Git Branching
##### git branch <branch-name> - Creates a new branch with the specified name.
##### git checkout <branch-name> - Switches to the specified branch.
##### git checkout -b <branch-name> - Creates a new branch and switches to it.
##### git merge <branch-name> - Merges the specified branch into the current branch.
##### git branch -d <branch-name> - Deletes the specified branch.
##### git branch -a - Lists all branches in the repository, including remote branches.
##### git branch -r - Lists all remote branches in the repository.
##### git push origin <branch-name> - Pushes the specified branch to the remote repository.
##### git pull origin <branch-name> - Pulls the latest changes from the specified branch in the remote repository to the local repository.
##### git clone <repository-url> - Clones a remote repository to the local machine.

### Repositories
##### A repository is a storage location for software packages, often used to store and manage code.
##### Repository - when a folder or file directory is monitored by Git is called a repository.
##### Many repositories are available on GitHub. Online Git Repositories - GitHub, GitLab, Bitbucket, SourceForge, AWS CodeCommit, Azure Repos, Google Cloud Source Repositories, and more.

### Deleting and Undoing changes
##### git rm <file> - Removes the specified file from the staging area and the working directory.
##### git reset <file> - Unstages the specified file, keeping the changes in the working directory.
##### git checkout -- <file> - Discards changes in the working directory and restores the specified file to the last committed state.
##### git revert <commit> - Creates a new commit that undoes the changes made in the specified commit.
##### git reset --hard <commit> - Resets the current branch to the specified commit, discarding all changes in the working directory and staging area.
##### git clean -f - Removes untracked files from the working directory.
##### git stash - Temporarily saves changes that are not ready to be committed, allowing you to switch branches or work on something else without losing your changes.  
##### git stash pop - Applies the most recently stashed changes and removes them from the stash list.
##### git stash list - Lists all stashed changes.
##### git stash drop <stash> - Removes the specified stash from the stash list.
```
