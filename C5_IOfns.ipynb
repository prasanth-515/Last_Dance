### Input and Output Functions

```python
### 1. Input functions

#### The functions that are used to take the input from the user are known as input functions.
#### It is used for dynamic initialisation
#### types of initialisation - 1. Static and 2. Dynamic

#### 1. Static Initialisation of values
#### initialisation the values within the code

a = 5  # static Initialisation
b = 6
c = a + b
print(c, type (c))
print()

#### 2. Dynamic Initialisation of values
#### code should ask the user to enter the value during run time

x = int (input("x = "))    # dynamic initialisation
y = float (input("y = "))
print('x = ',x)
print('y = ',y)
print()
z = x + y
a = x - y
b = x * y
if y!=0:
    c = x / y
else:
    c = "undefined (division by zero) "
print ("x + y = ", z)
print()
print ("x - y = ", a)
print()
print ("x * y =", b)
print()
print ("x / y = ", c)
```

```python
#### Input()

#### It is used to take some value from the user at run time.
#### It always returns a string value.
#### variable_name = input('Message') //syntax

x = input ("x = ")   # dynamic initialisation
y = input ("y = ")   # which will simply get altogether bcs it act as a strings
a = x + y
print(a)
print()

x = input ("x = ")   
print('x = ', x)              
y = input ("y = ")   
print('y = ', y)
a = int(x) + float (y)                                   # typecasting
print(f'The addition of {x} and {y} is {a}')             # f- string  
print('The addition of {} and {} is {}'.format(x,y,a))   # .format()
print()
```

```python
### 2. Output functions.

#### the functions that we used to print or show the output on the output screen(concole)
#### print() is an output function which prints the output from the concole
#### print(variable name)

print('Hello')  # output function

a = 10
b = 20
c = 30
d = 40
print(a,b,c,d)
print()
print(a,b,c,d, sep='-')
print(a,b,c,d, sep=' ')
print(a,b,c,d, sep='\n')
print()

print(a, end = ', ')
print(b, end = ', ')
print(c, end = ', ')
print (d)
```
