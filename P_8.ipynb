```python
# Use enumerate() on the list ['Ram','Kumar','Ravi'] and print the result as a list of tuples.

a = ['Ram','Kumar','Ravi']
e = enumerate(a,(1))
print(list(e))

# Use zip() to combine ['a','b','c'] with [1,2,3].

a =  zip(['a','b','c'], [1,2,3])
print(list(a))

# Use filter() with a lambda to get only even numbers from [1,2,3,4,5,6,7,8].

a = [1,2,3,4,5,6,7,8]
f = filter (lambda x:x%2==0, a)
print(list(f))

# Use map() with a lambda to square each number in [2,4,6,8].

a = [2,4,6,8]
m = map(lambda x:x**2,a)
print(list(m))
```
