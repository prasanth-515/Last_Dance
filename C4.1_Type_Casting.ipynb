### 2. Type Casting

```python
#### Type casting -  Converting type of one dat type to another data type. It is also called as type conversion.
#### eg: int to float, float to int, int to str, str to int, etc..

#### 1. int 
#### It typecaste to int. It returns an int value. We can convert a float, str, bool etc.. to int.
#### It is a whole number without a decimal point. It can be positive, negative, or zero. It is a built-in data type in Python.

a = 43
print(a, "is the value of a and its type is", type(a))
print()

b = 45.67
c = int(b)
print(c, "is the value of c and its type is", type(c))
print()

d = "247"
e = int(d)
print(e, "is the value of e and its type is", type(e))  
print()

f=True
g = int(f)
print(g, "is the value of g and its type is", type(g))
print()

#### 2. float
#### It typecaste to float. It returns a float value. We can convert an int, str, bool, etc.. to float.
#### It is a number with a decimal point. It can be positive, negative, or zero. It is a built-in data type in Python.
h = 43
print(h, "is the value of h and its type is", type(h))
print()

i = 45.67
j = float(i)
print(j, "is the value of j and its type is", type(j))
print()

k = "245"
l = float(k)
print(l, "is the value of l and its type is", type(l))
print()

m = False
n = float(m)
print(n, "is the value of n and its type is", type(n))
```

```python
#### 3. Complex()
#### It typecaste to complex. It returns a complex value. We can convert an int, float, str, bool, etc.. to complex.
#### It takes 2 numbers as arguments and returns a complex number.
#### complex(a,b) -  a is the real and b is the imaginary. a+bj.
#### if we giving only one argument, it is considered as real part and the imaginary part is 0. a+0j.

a =  complex(55,35)
print(a, "is the value of a and its type is", type(a))  
b = abs(a)
print(b, "absolute value of a", type(b))
c= a.conjugate()
print(c, "is the conjugate of a", type(c))
a = complex(-23.8)
print(a, type(a))
print(a.conjugate())
print(a.real)
print(a.imag)
print()

x=complex(True,False)
print(x, type(x))
print(x.conjugate())
print(x.real)
print(x.imag)
print()
```

```python
#### 4. str
#### It typecaste to str. It returns a string value. We can convert an int, float, bool, etc.. to str.
#### It is a sequence of characters. It is a built-in data type in Python.

a=43    
b=str(a)
print(b, "is the value of b and its type is", type(b))
print()

c=45.67
d=str(c)    
print(d, "is the value of d and its type is", type(d))
print()

e=True
f=str(e)    
print(f, "is the value of f and its type is", type(f))
print()

g=[1,2,3,4,5]
h=str(g)
print(h, "is the value of h and its type is", type(h))
print()
```

```python
#### 5. bool
#### It typecaste to bool. It returns a boolean value. We can convert an int, float, str, etc.. to bool.
#### It is a built-in data type in Python. It can have only two values: True and false.

a = 43
b = bool(a)
print(b, "is the value of b and its type is", type(b))
print()

c = 0.0
d = bool(c)
print(d, "is the value of d and its type is", type(d))
print()

e = "Prasanth"
print(len(e))
f = bool(e) 
print(f, "is the value of f and its type is", type(f))
print()

g = ""
print(len(g))
h = bool(g)
print(h, "is the value of h and its type is", type(h))
print()

i = [1,2,3,4,5]
j = bool(i)
print(j, "is the value of j and its type is", type(j))
print()

h = -3  
i = bool(h)
print(i, "is the value of i and its type is", type(i))
print()

x= None
y = bool(x)
print(y, "is the value of y and its type is", type(y))
print()
```

```python
#### 1. list
#### It typecaste to list. It returns a list value. We can convert a str, tuple, set, dict, etc.. to list.
#### It is a collection of items which is ordered and changeable.
#### if length function is working then its an iterable.

x = "Arithmetic Operations"
y= list(x)
print(y, type(y))
print(len(x))
print()

a = (1,2,3,4,5)
b = list(a)
print(b, type(b))
print()
```

```python
#### 2. tuple
#### It typecaste to tuple. It returns a tuple value. We can convert a str, list, set, dict, etc.. to tuple.
#### It is a collection of items which is ordered and unchangeable.

x = "Arithmetic Operations"
y = tuple(x)
print(y, type(y))
print()

a = [1,2,3,4,5]
b = tuple(a)
print(b, type(b))
print()
```

```python
#### 3. dict
#### It typecaste to dict. It returns a dict value. We can convert a str
#### A dictionary is a collection which is unordered, changeable and indexed. 
#### In Python dictionaries are written with curly brackets, and they have keys and values.
#### the iterable that we pass to the dict() must have one more iterable inside it.
#### the inner iterable must have two elements
#### we should pass containers of container

x = "Arithmetic Operations"
y = dict.fromkeys(x)
print(y, type(y))
print()

a = [1,2,3,4,5] 
b = (22)
c = dict.fromkeys(a, b)
print(c, type(c))
print()

d = {"Name": "Prasanth", "Age" : 30, "Sex" : "Male", "City" : "Salem"}
e = dict(d)
print(e, type(e))
print()

ke = list(d)
print(ke, type(ke))
print()

Va = list(d.values())
print(Va, type(Va))

print()

x= ([3,"Prasanth"],(30,"Age"),{4,'Place'})
y=dict(x)
print(y, type(y))
print()
z = dict.fromkeys(y)
print(z)

```

```python
#### 4. set
#### It typecaste to set. It returns a set value. We can convert a str, list, tuple, dict, etc.. to set.
#### A set is a collection which is unordered, unchangeable*, and unindexed. In Python sets are written with curly brackets.

x = "Arithmetic Operations"
y = set(x)
print(y, type(y))
print()

a = [1,2,3,4,5]
b = set(a)
print(b, type(b))
print()

c = (1,2,3,3,6,4,5)
d = set(c)
print(d, type(d))
print()

e = {1: "one", 2: "two", 3: "three", 4: "four", 5: "five"}
f = set(e)
print(f, type(f))
print()

```

```python
#### 5. frozenset
#### It typecaste to frozenset. It returns a frozenset value. We can convert a str, list, tuple, dict, etc.. to frozenset.
#### A frozenset is a collection which is unordered, unchangeable, and unindexed. In Python frozensets are written with curly brackets.

x = "Arithmetic Operations"
y = frozenset(x)
print(y, type(y))
print()

a = [1,2,3,4,5]
b = frozenset(a)
print(b, type(b))
print()

c = (1,2,3,3,6,4,5)
d = frozenset(c)
print(d, type(d))
print()

e = {1: "one", 2: "two", 3: "three", 4: "four", 5: "five"}
f = frozenset(e)
g = frozenset(e.values())
h = frozenset(dict.fromkeys(e))
print(h,type(h))
print(g,type(g))
print(f, type(f))
print()

g = frozenset({1: "one", 2: "two", 3: "three", 4: "four", 5: "five"})
print(g, type(g))
print()

```

```python
##### 6. range
#### It typecaste to range. It returns a range value. We can convert an int, float, str, etc.. to range.
#### A range is a sequence of numbers, starting from 0 by default, and increments by 1 (by default), and ends at a specified number. 
#### In Python, the range() function is used to generate a sequence of numbers.

x = 10
y = range(x)    
print(list(y), type(y))
print()

a = 5.67
b = range(int(a))
print(list(b), type(b))
print()

c = "Python"
d = range(len(c))
print(list(d), type(d))
print()

e ="25"
f = range(int(e))
print(list(f), type(f))
print()
```
