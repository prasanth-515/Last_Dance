```python
# - A function can be used within the program that it ahs been defined in.
# - we can not use it outside the program.

# Moudules
# - A module is a collection of functions.
# - We can use functions from a module in any program by importing the module.
```

```python
# Types of modules.
# 1. Built-in modules.
# 2. User defined moudules.
```

```python
# 1. Built-in modules.

# - The moudules that are predefined in python are the builtin modules.
# - ex: math, sys, os, functools, openai, collections etc.
# - To use any module we have to import it first.

# Importing a module.

# Case 1:

# import module_name

import math  #importing math module.
print(dir())

# - . operator is used to invoke any function inside the module.
# - moudule_name.function_name()

print(f'square root {int(math.sqrt(16))}')
print(f'factorial {int(math.factorial(5))}')
print()

help(math.sqrt) # which will give the details of the module
```

```python
# Case 2:
# - Importing a specific function (function) from a module.

# from moudle_name import function_name, function name....

print(dir())
print()

from math import factorial, sqrt, pi
print(dir())
```

```python
# Case 3:

# - Importing all the functions from a module.

# from module_name import *

from math import *
print(dir())
```

```python
# Case 4:

# - Renaming a module.

# import module_name as given_name

import math as mt
print(dir())
print()
print(int(mt.sqrt(4)))
```
