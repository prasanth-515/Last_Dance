```python
# Q1. Print numbers from 1 to N
# Enter N: 5
# 1 2 3 4 5

n = int(input('Enter a number : '))
print('Enter a number : ',n)
print()

for i in range(1,n+1):
    print(i,end=' ')
```

```python
# Q2. Find sum of all numbers from 1 to N
# Enter N: 5
# 1+2+3+4+5 = 15

n = int(input('Enter a number : '))
print('Enter a number : ',n)
print()
s = 0
for i in range(1,n+1):
    s = s + i
    if i < n:
        print(f'{i} + ', end='')
    else:
        print(f'{i}', end='')
print(f' = {s}')
```

```python
# Q3. Print multiplication table of a number
# Enter number: 3
# 3 x 1 = 3
# 3 x 2 = 6
# ...
# 3 x 10 = 30

a = int(input('Enter a Mulitplicand : '))
print('Enter a Multiplicand : ',a)
b = int(input('Enter a Multiplier : '))
print('Enter a Multiplier : ',b)
print()

for i in range(1,b+1):
    c = a * i
    print(f'{a} x {b} = {c}')
```

```python
# Q4. Count vowels in a string
# Enter string: Prasanth
# Vowels: a a
# Vowel count = 2

s = str(input('Enter a string : '))
print('Enter a string : ',s)
print()
v = ''
c = 0
for i in range(len(s)):
    if s[i] in 'aeiouAEIOU':
        c += 1
        v += s[i] + ' '
print(f'Vowels: {v}')
print('Vowels Count : ',c)
```

```python
# Q5. Find largest number in a list
# numbers = [3, 67, 23, 89, 12, 45]
# Largest number = 89

n = [3, 67, 23, 89, 12, 45]
print(f'n = {n}')
l = 0
for i in range(len(n)):
    if n[i] > l:
        l = n[i]
print('Larges number = ',l)
```
