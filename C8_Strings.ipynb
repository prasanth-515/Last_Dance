```python
# Captilize() method

# - If the first letter is an alphabet in a string, it capitalises that.
# - It retruns a new object. It does not make the changes in the existing string. 

a = 'hello world'
print(a, type(a), a.capitalize()) # The capitalized one saved in new memory location and the original string is unchanged. 
```

```python
# Title() method

# - It capitalises the first letter of each word in a string.

a = 'i am the way, truth and the life'
print(a, '( <- title() -> )', a.title()) # The title one saved in new memory location and the original string is unchanged.
```

```python
# swapcase() method
# - It converts the uppercase letters to lowercase and lowercase letters to uppercase in a string.

a = 'i am the way, truth and the life'
print(a, '( <- swapcase() -> )', a.swapcase()) # The swapcase one saved in new memory location and the original string is unchanged.
print()

b = 'HELLO WORLD'
print(b, '( <- swapcase() -> )', b.swapcase()) # The swapcase one saved in new memory location and the original string is unchanged.
```

```python
# upper() method
# - It converts all the lowercase letters to uppercase in a string. 

a = 'i am the way, truth and the life'
print(a, '( <- upper() -> )', a.upper())
```

```python
# lower() method
# - It converts all the uppercase letters to lowercase in a string.

a = 'LOVE ONE ANOTHER'
print(a, '( <- lower() -> )', a.lower())    
```

```python
#  strip() method
# - It removes the specified character from both the boundaries of the string.
# - By default, it removes space characters.
# - It retruns a new object. It does not make the changes in the existing string.

a = '###He#llo World###'  #only remove from the boundaries.
print(a, '( <- strip() -> )', a.strip('#')) # The stripped one saved in new memory location and the original string is unchanged.
print()

b = '               Hello World                   '
print(b, '( <- strip() -> )', b.strip()) # remove the empyt spaces.
```

```python
# lstrip() method
# - It removes the specified character from the left boundary of the string.
# - By default, it removes space characters.

a = '$ 150 $'  #only remove from the left boundary.
print(a, '( <- lstrip() -> )', a.lstrip('$')) 
print()

# rstrip() method
# - It removes the specified character from the right boundary of the string.   
# - By default, it removes space characters.

b = '###Hello World###'  #only remove from the right boundary.
print(b, '( <- rstrip() -> )', b.rstrip('#'))
```

```python
# split() method
# - It splits a string into a list of substrings based on a specified separator.
# - By default, it splits the string based on whitespace characters.

a = 'Hello#World#Python#Programming'
print(a, '( <- split() -> )', a.split('#')) # The splitted one saved in new memory location and the original string is unchanged.
b = a.split('#')
print(b, type(b))
print(len(b))
print()

c = tuple(b)
print(c, type(c))
print()
```

```python
# splitlines() method
# - It splits a string into a list of lines based on the newline character ('\n').

a = 'Hello World\nPython Programming\nWelcome to the world of strings'
print(a, '( <- splitlines() -> )', a.splitlines(keepends=True))
print()
```

```python
# join() method
# - It takes a list or tuple of strings and return a string. 
# - It combines the strings of the list separated by a specified character and returns a string. 

a = ['Love', 'One', 'Another', 'As', 'I', 'Have', 'Loved', 'You']
print(a, '( <- join() -> )', '_'.join(a)) # the string which we passing before the join(), will act as a separator.
print()

b = ('\n\nHello', 'World', 'Python', 'Programming')
print(b,'( <- join() -> )', '\n'.join(b)) # the string which we passing before the join(), will act as a separator.
```

```python
# index() method
# - It returns the index of the first occurrence from left to right of the specified character. 
# - If the character is not found, it raises a ValueError.
# a.index('character')
# a.index('character', start_index, stop_index)

a = ' I am the way, truth and the life '
print(a.index('t')) # it will return the index of the first occurrence of 't' from left to right.
print(a.index('t', 10)) # it will return the index of the first occurrence of 't' from left to right after index 10.
print(a.index('t', 16, 29)) # it will return the index of the first occurrence of 't' from left to right between indices 10 and 20.
# print(a.index('z')) # it will raise a ValueError since 'z' is not found in the string.
print()
```

```python
# find() method
# - It returns the index of the first occurrence from left to right of the specified character. 
# - If the character is not found, it returns -1.   

a = ' I am the way, truth and the life '
print(a.find('t')) # it will return the index of the first occurrence of 't' from left to right.
print(a.find('t', 10)) # it will return the index of the first occurrence of 't' from left to right after index 10.
print(a.find('t', 16, 29)) # it will return the index of the first occurrence of 't' from left to right between indices 10 and 20.
print(a.find('z')) # it will return -1 since 'z' is not found in the string.
print()
```

```python
# rfind() method

# - It returns the index of the first occurrence from right to left of the specified character.
# - If the character is not found, it returns -1.

a = ' I am the way, truth and the life. Jesus is the way, truth and the life. '
print(a.rfind('t')) # it will return the index of the first occurrence of 't' from right to left.
print(a.rfind('t', 10)) # it will return the index of the first occurrence of 't' from right to left after index 10.
print(a.rfind('t', 16, 29)) # it will return the index of the first occurrence of 't' from right to left between indices 10 and 20.
print(a.rfind('z')) # it will return -1 since 'z' is not found in the string.   
print()
```

```python
# count() method
# - It returns the number of occurrences of a specified character in a string.
# - if the substring is not found, it returns 0.


a = ' I am the way, truth and the life. Jesus is the way, truth and the life. '
print(a.count('t')) # it will return the number of occurrences of 't' in the string.
print(a.count('t', 10)) # it will return the number of occurrences of 't' after index 10 in the string.
print(a.count('t', 16, 29)) # it will return the number of occurrences of 't' between indices 10 and 20 in the string.
print(a.count('z')) # it will return 0 since 'z' is not found in the string.

```

```python
# replace() method
# - It replaces all occurrences of a specified character with another character in a string.

a = ' I am the way, truth and the life. Jesus is the way, truth and the life. '
print(a.replace('truth', 'TRUTH')) # it will replace all occurrences of 't'
```

```python
# isalnum() method
# - It returns True if all characters in the string are alphanumeric (letters and numbers) and 
# - there is at least one character, otherwise it returns False. 

a = 'HelloWorld123'
print(a.isalnum()) # it will return True since all characters in the string are alphanumeric and there is at least one character.
print()

b = 'Hello World 123'
print(b.isalnum()) # it will return False since there are space characters in the string
```

```python
# isalpha() method
# - It returns True if all characters in the string are alphabets, otherwise it returns False.

a = 'HelloWorld'
print(a.isalpha()) # it will return True since all characters in the string are alphabets and there is at least one character.
print()

b = 'Hello World'
print(b.isalpha()) # it will return False since there are space characters in the string    
```

```python
# istitle() method
# - It returns True if the string is in title case (i.e., the first letter
# of each word is capitalized and the rest are lowercase), otherwise it returns False.

a = 'Hello World'
print(a.istitle()) # it will return True since the string is in title case. 
print()
```

```python
# isupper() method
# - It returns True if all characters in the string are uppercase, otherwise it returns False.

a = 'HELLO WORLD'
print(a.isupper()) # it will return True since all characters in the string are uppercase.  

b = 'HELLO wORLD'
print(b.isupper()) # it will return False since there are lowercase characters in the string.

```

```python
# islower() method
# - It returns True if all characters in the string are lowercase, otherwise it returns False.

a = 'hello world'
print(a.islower()) # it will return True since all characters in the string are lowercase.

B = 'Hello World'
print(B.islower()) # it will return False since the first character is uppercase.
```

```python
# isidentifier() method
# - It returns True if the string is a valid identifier (i.e., it can be used as a variable name), otherwise it returns False.

# properties of identifiers.
# 1. It can only contain letters (both uppercase and lowercase), digits, and underscores (_).
# 2. It must start with a letter (both uppercase and lowercase) or an underscore (_). It cannot start with a digit.
# 3. It cannot be a reserved keyword in Python (e.g., if, else, for, while, etc.).


a = 'variable_name'
print('a is', a.isidentifier()) # it will return True since the string is a valid identifier.

b = '1variable_name'
print('b is', b.isidentifier()) # it will return False since the string starts with a number.

c = '_variable_name22'
print('c is', c.isidentifier()) # it will return True since the string starts with an underscore and contains only valid characters.

d = 'variable-name'
print('d is', d.isidentifier()) # it will return False since the string contains a hyphen which is  not a valid character for an identifier.
```

```python
# isnumeric() method, isdigit() method, isdecimal() method

# isnumeric() method
# - It returns True if the string contains only numeric characters (0-9), otherwise it returns False.
a= '12345'
print('a(12345) is', a.isnumeric()) # it will return True since all characters in the string are numeric characters
print()
x = '12345.67'
print('x(12345.67) is', x.isnumeric()) # it will return False since there is a decimal point in the string.
print()

# isdigit() method
# - It returns True if the string contains only digits (0-9), otherwise it returns False.
b = '12345'
print('b(12345) is', b.isdigit()) # it will return True since all characters in the string are digits
print()
y = '12345.67'
print('y(12345.67) is', y.isdigit()) # it will return False since there is a decimal point in the string.
print()

# isdecimal() method
# - It returns True if the string contains only decimal characters (0-9), otherwise it returns False.
c = '12345' 
print('c(12345) is', c.isdecimal()) # it will return True since all characters in the string are decimal characters
print()
z = '12345.67'
print('z(12345.67) is', z.isdecimal()) # it will return False since there is a decimal point in the string.

```

```python
# startswith() method
# - It returns True if the string starts with the specified character, otherwise it returns False.

a = 'Hello World'
print('a(Hello World startswith("H")) is ', a.startswith('H')) # it will return True since the string starts with 'H'.
print()

b = 'Hello World'
print('b(Hello World startswith("h")) is ', b.startswith('h')) # it will return False since the string does not start with 'h'.
print()

# endswith() method
# - It returns True if the string ends with the specified character, otherwise it returns False.

c = 'Hello World'
print('c(Hello World endswith("d")) is ', c.endswith('d')) # it will return True since the string ends with 'd'.
print()

d = 'Hello World'
print('d(Hello World endswith("D")) is ', d.endswith('D')) # it will return False since the string does not end with 'D'.
print()

e = 'Hello World'
print('e(Hello World endswith("World")) is ', e.endswith('dlrow')) #
```

```python
# Membership operation.

# A string supports membership operation. 
# in and not in are the membership operators in Python.


a = 'Hello World'
print('a(Hello World in "Hello")) is ','Hello' in a) # it will return True since 'Hello' is a substring of 'Hello World'.
print()

b = 'Hello World'
print('b(Hello World in "world")) is ', b in 'world') # it will return False since 'world' is not a substring of 'Hello World'.
print()

c = 'Hello World'
print('c(Hello World not in "Hello")) is ', c not in 'Hello') # it will return False since 'Hello' is a substring of 'Hello World'.
print()

d = 'Hello World'
print('d(Hello World not in "world")) is ', d not in 'world') # it will return True since 'world' is not a substring of 'Hello World'.
print()
```
