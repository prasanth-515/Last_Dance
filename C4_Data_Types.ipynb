### 1. Data types

```python
### 1. Data types
##### it is a classifier that classifies the type of a value. It is a built-in function in python.
##### it is used to check the type of a value or variable.
##### different type of data need different type of operation and memory allocation. 

#### types of data types:
#### 1. Premitive data types: int, float, bool, str, etc..
#### 2. Non-primitive data types: list, tuple, set, dict, etc
```

#### 1. Primitive data types

```python
#### 1. Integer (int): it is a whole number without a decimal point. It can be positive, negative, or zero.
a = 0
print("Integer (int) ")
print(a, "is an integer", type(a))
print()

#### 2. Float (float): it is a number with a decimal point. It can be positive, negative, or zero.
b = 0.0
print("Float (float) ")
print(b, "is a float", type(b))
print()

#### 3. complex (complex): it is a number with a real part and an imaginary part. 
#### It is represented as a + bj, where a is the real part and b is the imaginary part.
c = 2 + 3j
print("Complex (complex) ")
print(c, "is a complex number", type(c))  
print(c.real, "is the real part of c")
print(c.imag, "is the imaginary part of c") 
print (c.conjugate(), "is the conjugate of c")
print (abs(c), "is the magnitude of c")
print ((2**2+3**2)**0.5, "is the magnitude of c calculated using the formula")
print()

```

```python
#### 4. string (str): it is a sequence of characters enclosed in single quotes, double quotes, or triple quotes.
print("String (str) ")
d = "Hello, World!"
print(d, "is a string", type(d))
print(d[0], "is the first character of d")
print(d[6], "is the seventh character of d")
print()

#### 5. boolean (bool): it is a data type that can have only two values: True or False. 
#### It is used to represent the truth value of an expression, case sensitive.
print("Boolean (bool) ")
e = True
f = False
print(e, "is a boolean", type(e))
print(f, "is a boolean", type(f))   
print(e and f, "is the result of e and f")
print(e or f, "is the result of e or f")
print(not e, "is the result of not e")
print(not f, "is the result of not f")
print()


#### 6. NoneType (None): it is a data type that has only one value: None. 
#### It is used to represent the absence or missing value or a null value.
print("NoneType (None) ")
g = None
print(g, "is a NoneType", type(g))
name = 'Prasanth'
print(name, "is a string", type(name))
age = None
print(age, "is a NoneType", type(age))
print()
```

#### 2. Derived data types

```python
### 2. Derived data types
#### it is a data type that is derived from the primitive data types. 
#### These are data structures which is used to store multiple values in a single variable.
```

```python
#### 1. List (list): it is a collection of items which is ordered and changeable. List is mutable.
#### It is defined by enclosing the items in square brackets [] and separating them with commas.
#### It is heterogeneous, which means it can store different types of data in a single list.
#### it may have duplicate values, which means it can store the same value multiple times in a list.
print("List (list) ")
h = [1, 2, 3, 4, 5]
i = ["Hello", "World", 1, 2, 3]
print(h, "is a list", type(h))
print(h[0], "is the first element of h")
print(h[4], "is the fifth element of h")
print(i, "is a list", type(i))
print(i[0], "is the first element of i")
print(i[4], "is the fifth element of i")
print()

#### 2. Tuple (tuple): it is a collection of items which is ordered and unchangeable. Tuple is immutable.
#### It is defined by enclosing the items in parentheses () and separating them with commas.
#### It is heterogeneous, which means it can store different types of data in a single tuple.
#### it may have duplicate values, which means it can store the same value multiple times in a tuple.
print("Tuple (tuple) ")
j = (1, 2, 3, 4, 5)
k = ("Hello", "World", 1, 2, 3.3, True, None)
print(j, "is a tuple", type(j))
print(j[0], "is the first element of j")
print(j[4], "is the fifth element of j")
print(k, "is a tuple", type(k))
print(k[0], "is the first element of k")
print(k[6], "is the seventh element of k")
print()


### List vs Tuple
#### List is defined by enclosing the items in square brackets [] and separating them with commas.
#### Tuple is defined by enclosing the items in parentheses () and separating them with commas.
#### List is mutable, which means it can be changed after it is created.
#### Tuple is immutable, which means it cannot be changed after it is created.

#### mutabiliy vs immutability
print("Mutability vs Immutability")
lis = [22, 43, 54, 65, 76]
tup = (54, 94, 34, 23, 12)

lis[0] = 100
print(lis, "is mutable")

# tup[0] = 100 - it gives an error because tuple is immutable.
print(tup, "is immutable")

```

```python
#### 3. Dictionary (dict): it is a collection of items which is unordered, changeable, and indexed. Mutable.
#### Enclosed with curly braces {} and separating them with commas.   
#### it is in the form of key-value pairs. Values are mapped with keys.
#### duplicate values are allowed.


print("Dictionary (dict) ")
n = {"name": "Prasanth", "age": 22, "city": "Chennai"}
print(n, type(n))
print(n["name"])
print(n["age"], "is the value of the key 'age' in n")
print(n["city"], "is the value of the key 'city' in n")
n["age"] = 23
print(n, "is the updated dictionary after changing the value of the key 'age'")
print()

```

```python
#### 4. Set (set): it is a collection of items which is unordered, unchangeable, and unindexed.
#### Collection of hashable items Enclosed with curly braces {} and separating them with commas.
#### Hashable means that the items in a set must be immutable, which means they cannot be changed after they are created.
#### hasable items are int, float, bool, str, tuple, etc.. because they are immutable.
#### unhashable items are list, dict, set, etc.. because they are mutable.
print("Set (set) ")
# o = { (1,3,5), [2,4,6] } # it gives an error because list is unhashable.
s = { 1, 2, (32, "Hello"), 3.3, True, None, 3.3 }
print(s, type(s)) 
print()


#### 5. Frozenset (frozenset): it is a collection of items which is unordered, unchangeable, and unindexed. Immutable.
#### Collection of hashable items Enclosed with curly braces {} and separating them with commas.
#### An iterable inside the frozenset() function is used to create a frozenset.
#### Iterable means that it can be looped through, which means it can be used in a for loop or a while loop.
#### the iterable that we pass inside the frozenset() must be hashable.
#### inside the frozenset there should be iterable, inside the iterable there should be only hashable items. 
#### iterables are list, tuple, set, dict, string, int, float, bool, none, etc..

print("Frozenset (frozenset) ")
fro = frozenset({1, 2, (32, "Hello"), 3.3, True, None, 3.3})
print(fro, type(fro)) 
print()

xyz = frozenset({2,3,("hello", 4.5, 2+2j), "Prasanth", None, True})
for i in xyz:
 print(i, type(xyz))
```

```python
#### Set vs Frozenset

#### set - index are not allowed because it is unordered and unindexed. We can add or remove elements in set.
#### frozenset - index are not allowed because it is unordered and unindexed. We cannot add or remove elements in frozenset.
#### set is mutable whereas frozenset is immutable.
set1 = {1, 2, 3, 4, 5}
fro1 = frozenset({1, 2, 3, 4, 5})
print(set1, "is a set", type(set1))
print(fro1, "is a frozenset", type(fro1))
set1.add(6)
set1.remove(2)
print(set1, "is a set after adding an element")
# fro1.add(6) - it gives an error because frozenset is immutable.
print(fro1, "is a frozenset after trying to add an element")
print()
```

```python
#### 6. Range (range): it is a collection of items which is ordered, unchangeable, and indexed. Immutable.
#### It is defined by the range() function, which takes three parameters: start, stop, and step.
#### start is the starting point of the range, which is inclusive. It is optional and default is 0.
#### stop is the ending point of the range, which is exclusive. It is mandatory.
#### if we pass only one value in the range() function, it is considered as stop and start is considered as 0.
#### step is the increment or decrement value of the range. It is optional and default is 1.
#### float values are not allowed in the range() function because it is used to generate a sequence of integers.

print("Range (range) ")
r = range(0, 11, 1)
print(r, type(r))

ra = list(r)
print (ra)
print()

x = range(3,31,3)
y = list(x)
print(y)
print()

z = list(range(1,21,2))
print(z, 'Multiples of 2 till 20')
print()

print(list(range(10,-1,-1)), 'Countdown from 10 to 0')
print()

print(list(range(5)), 'Range with only stop value')
print()

print(list(range(1,10,-1)), 'Range with negative step value')
print()

print(list(range(5,1,1)), 'Range with positive step value but start is greater than stop')
```
