#### If and If else

```python
# Syntax.
# if condition:
#     statement(s) // when the if condition is true then the code inside the if block will be executed. 

if 2==2:
    print('Hello') # it will print Hello since the condition is true.

if 2==3:
    print('Haaaiii') # it will not print Hello since the condition is false.
print()

# if else statement.
# - It is used to execute a block of code when the condition is true and another block of code when the condition is false.

a = int(input('Enter a number: '))
if a>3:
    print('Final_Dance')
else:
    print('Last_Dance')
print()

x = int(input('Enter the value of x = '))
print('x = ',x)
y= int(input('Enter the value of y = '))
print('y = ',y)
if x>y:
    print(f'x({x}) is greater than y({y})')
else:
    print('y({}) is greater than x({})'.format(y,x))

```

```python
# Multiple if statements.

# sytnax.
# if condition1:
#     statement(s) # when the condition1 is true then the code inside the if block will be executed.
# if condition2:
#     statement(s) # when the condition2 is true then the code inside the if block will be executed.
# if condition3:
#     statement(s) # when the condition3 is true then the code inside the if block will be executed.

# 1. In the above condition, all the conditions are checked one by one from top to bottom.
# 2. If all the conditions are true then all the statements inside the if blocks will be executed.
# 3. If any one of the condition is false then the statements inside that if block

a = int(input('Enter a number: '))
if a == 1:
    print(f'{a} is one')
if a == 2:
    print(f'{a} is two')
if a == 3:
    print(f'{a} is three')
if a == 4:
    print(f'{a} is four')
if a == 5:
    print(f'{a} is five')
else:
    print(f'a({a}) enter a valid number')  # this else is only related to the previous if condition.

# if-elif-else statement.
# - It is used to execute a block of code when the condition is true and another block
# of code when the condition is false. It is used when we have multiple conditions to check.


a = int(input('Enter a number: '))  # elif is faster than if else condition
if a == 1:
    print(f'{a} is one')            # stop the execution at the condition where it is true, so it won't run further.
elif a == 2:
    print(f'{a} is two')            
elif a == 3:
    print(f'{a} is three')
elif a == 4:
    print(f'{a} is four')
elif a == 5:
    print(f'{a} is five')
elif a < 0:
    print(f'{a} is negagtive number')
else:
    print(f'a({a}) enter a valid number') # this else is related for all the elif condition.
```

```python
# # Nested if else statement.
# # whenever we have an if statement within another if statement, we call it as nested if statement.

# # syntax
# # if condition:
#     if condition:
#         Code
#     else:
#         Code
#   else:
#     code


Name = str(input('Enter your name '))
print(f'Name - {Name}')
a = int(input('Enter a = '))

if a in range(0,100):
    print('Marks - {}'.format(a))
    if a >= 90:
        print("Grade - A")
    elif a >= 80:
        print('Grade - B')
    elif a >= 70:
        print('Grade - C')
    elif a >= 60:
        print('Grade - D')
    elif a in range (0,60):
        print('Fail - Apply for revaluation')
else:
    print('Enter  a valid Mark')
```

```python
# Ternary Operator
# It is a one liner code to write a conditional statement.
# It is a substitute to if...else statement.

# syntax exp_1 if condition else exp_2
# 1. First the if condition is checked.
# 2. Suppose the if condition id True then exp_1 will be executed.
# 3. Suppose if condition is False then exp_2 will get executed. 

# print('A') if 2==2 else print('B')

a = int(input('Enter a : '))
print('Enter the value of a : ', a)
b = int(input('Enter b : ')) 
print('Enter the value of b : ', b)
c = int(input('Enter c : '))
print('Enter the value of c : ', c)
print()

print(f'A({a}) is greater') if a>b and a>c else print(f'B({b}) is greater') if b>a and b>c else print('C({}) is greater'.format(c))
```
