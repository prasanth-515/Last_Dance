```python
# 1️⃣ CONDITIONAL STATEMENTS (if-elif-else)
# Q1. Interview Classic: Check if a year is a Leap Year
# Enter year: 2024
# 2024 is a Leap Year ✅

x = int(input('Enter a year : '))
print('Enter a year : ',x)
print()

if x % 4 == 0 and (x % 100 != 0 or x % 400 == 0):
    print(f'{x} is a Leap year') 
else:
    print(f'{x} is not a Leap year')
```

```python
# Q2. Real World: Calculate Electricity Bill based on units consumed
# 0-100 units: ₹5/unit
# 101-200 units: ₹7/unit
# 201+ units: ₹10/unit

x = int(input("Enter your units : "))
print('Enter your units : ',x)
print()

if x >= 0 and x <= 100:
    c = x * 5
    print(f'Electricity Bill : ₹ {c}')
elif x >= 101 and x <= 200:
    d = (100 * 5) + (x - 100) * 7
    print(f'Electricity Bill : ₹ {d}')
elif x >= 201:
    e = (100 * 5) + (100 * 7) + (x - 200) * 10
    print(f'Electricity Bill : ₹ {e}')
else:
    print('Enter a valid unit')
```

```python
# Q3. Interview: Find the largest of 3 numbers without using max()

a = int(input('Enter a : '))
print('Enter a : ',a)
b = int(input('Enter b : '))
print('Enter b : ',b)
c = int(input('Enter c : '))
print('Enter c : ',c)
print()

if a >= b and a >= c:
    print(f'{a} is largest ')
elif b >= a and b >= c:
    print(f'{b} is largest ')
else:
    print(f'{c} is largest')
```

```python
# Q4. Real World: BMI Calculator with health category
# BMI < 18.5: Underweight
# 18.5-24.9: Normal
# 25-29.9: Overweight
# 30+: Obese

w = float(input('Enter w : '))
print(f'Enter your weight : {w} (kg)')
h = float(input('Enter h : '))
print('Enter your Height : {} (m)'.format(h))
print()

BMI = w / (h*h)
print('Body Mass Index : ',(round(BMI,2)))
print()

if BMI < 18.5:
    print('YOu are Underweight')
elif BMI >= 18.5 and BMI <= 24.9:
    print('You are normal')
elif BMI >= 25 and BMI <= 29.9:
    print('You are Overweight')
elif BMI >= 30:
    print('You are Obese')
else:
    print('Recheck you Height and weight \n Enter a valid Height and weight')
```

```python
# Q5. Learning: Check if a triangle is valid and classify it (Equilateral/Isosceles/Scalene)

a = int(input('Enter a : '))
print('Enter a : ',a)
b = int(input('Enter b : '))
print('Enter b : ',b)
c = int(input('Enter c : '))
print('Enter c : ',c)
print()

if a + b > c and a + c > b and b + c > a:
    print('The Triangle is Valid')
    if a == b == c:
        print('It is equilateral Triangle --> "All Sides are Equal"')
    elif a == b or b == c or c == a:
        print('It is Isosceles Triangle --> "Two Sides are Equal"')
    else:
        print('It is Scalene Triangle --> "All sides are not equal"')
else:
    print('It is not a valid Triangle')
```

```python
# Q6. Interview: FizzBuzz problem (classic in EVERY interview!)
# If divisible by 3 → "Fizz"
# If divisible by 5 → "Buzz"  
# If divisible by both → "FizzBuzz"
# Else → print the number

x = int(input('Enter x : '))
print('Enter x : ',x)
print()

if x % 3 == 0 or x % 5 == 0:
    if x % 3 == 0 and x % 5 != 0:
        print('Fizz')
    elif x % 5 == 0 and x % 3 != 0:
        print('Buzz')
    else:
        print('FizzBuzz')
else:
    print('Print the number : ',x)
```

```python
# Q7. Real World: Income Tax Calculator (Indian Slabs)
# 0-2.5L: No tax
# 2.5-5L: 5%
# 5-10L: 20%
# 10L+: 30%

x = float(input('Enter x : '))
print(f'Enter your Salary : ₹ {x} /-')
print()

if x <= 250000:
    print('You are Tax free')
    print('Your Salary is : ',x)
elif x > 250000 and x <= 500000:
    a = (x - 250000) * 5 / 100
    print('Your Tax is : ',round(a,2))
    print()
    a1 = x - a
    print('Your Salary after Tax : ',a1)
elif x > 500000 and x <= 1000000:
    b = ( (250000) * 5 / 100 )+ ( ( x - 500000) * 20 / 100) 
    print(' Your Tax is : ', round(b,2))
    print()
    b1 = x - b
    print('Your salary after Tax : ',b1)
else:
    c = ( (250000) * 5 / 100 ) + ( (500000) * 20 / 100) + ( ( x - 1000000) * 30 / 100)
    print(' Your Tax is : ', round(c,2))
    print()
    c1 = x - c
    print('Your salary after Tax : ',c1)
```
