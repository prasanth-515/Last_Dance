```python
## Shortcut keys

#### - " Esc " to enter into command mode
#### - " Enter " to enter into edit mode    
#### - " A " to add a cell above the current cell
#### - " B " to add a cell below the current cell
#### - " D D " to delete the current cell
#### - " M " to change the cell type to markdown    
#### - " Y " to change the cell type to code
#### - " Shift + Enter " to run the current cell and move to the next cell
#### - " Ctrl + Enter " to run the current cell and stay in the same cell
#### - " Alt + Enter " to run the current cell and insert a new cell below
#### - " Ctrl + S " to save the notebook
#### - " Ctrl + Z " to undo the last action
#### - " Ctrl + Shift + Z " to redo the last action
#### - " Ctrl + C " to copy the current cell    
#### - " X " to delete the current cell
#### - " Ctrl + V " to paste the copied cell
#### - " z " to undo the last action
```

```python
## Programming Language

#### It is a language used to communicate with the machine.
#### It is a set of instructions that the machine can understand and execute.
#### Some of the popular programming languages are : Python, Java, C++, etc.    
#### Programming languages is a set of some reserved words and some syntax(rules) which is used to write the code.

## Types of Programming Languages
#### - High Level Language - Human readable language. It is easy to understand and write. 
#### - Low Level Language. It is machine readable language. It is difficult to understand and write.
#### - Assembly Language. It is a low level language which is used to write the code for the hardware - Assembly code.
#### - Machine Language. It is a low level language which is used to write the code for the hardware - Binary code.
#### - Scripting Language. It is a high level language which is used to write the code for the software - Python, JavaScript, etc.
#### - Markup Language. It is a language which is used to write the code for the web pages - HTML, XML, etc.
```
