## 8. Bitwise Operators

##### These operators are used to perform bit level operations.
##### A binary digit is known as a **bit**.

### Steps to perform Bitwise Operations:
##### 1. Convert the number into binary digits (bits)
##### 2. Perform the bitwise operation bit by bit
##### 3. Convert the result back to a number

### Note:
##### Bitwise operators are only applicable on **int** values
##### Because only int values can be converted to binary digits

### List of Bitwise Operators:
##### `&`  → Bitwise AND
##### `|`  → Bitwise OR
##### `~`  → Bitwise NOT
##### `^`  → Bitwise XOR
##### `<<` → Bitwise Left Shift
##### `>>` → Bitwise Right Shift

### 1. Bitwise AND - `&`

##### Logic : `1 AND 1 = 1`, else `0`
##### (Same as logical AND but at bit level)

```python
a = 78
b = 235

print('bin(a) =', bin(a))   # 0b01001110
print('bin(b) =', bin(b))   # 0b11101011

c = a & b
print('a & b =', c, bin(c)) # 0b01001010
print()

# Adding 0b before a binary number gives the integer of that binary number
print('0b01001010 =', 0b01001010)
```

### 2. Bitwise OR - `|`

##### Logic : `0 OR 0 = 0`, else `1`
##### (Same as logical OR but at bit level)

```python
a = 182
b = 392

print('bin(a) =', bin(a))   # 0b10110110
print('bin(b) =', bin(b))   # 0b110001000

c = a | b
print('a | b =', c, bin(c)) # 0b110111110
print()

print('0b110111110 =', 0b110111110)
```

### 3. Bitwise NOT - `~`

##### Formula : `~num = -(num + 1)`
##### It flips all the bits of the number

```python
a = 5
b = ~a

print('~ of 5 is', b)  # -(5+1) = -6
```

### 4. Bitwise XOR - `^` (Caret Sign)

##### Logic : Same digits return `0`, opposite digits return `1`

```python
a = 182
b = 392

print('bin(a) =', bin(a))   # 0b10110110
print('bin(b) =', bin(b))   # 0b110001000

c = a ^ b
print('a ^ b =', c, bin(c)) # 0b100111110
print()

print('0b100111110 =', 0b100111110)
```

### 5. Bitwise Left Shift - `<<`

##### It shifts the binary digits to the **left** by the specified number of bits
##### Syntax : `number << number_of_bits`

### Steps:
##### 1. Convert the number to binary digits
##### 2. Represent the number in a 32-bit system
##### 3. Remove the specified number of bits from the **left** side
##### 4. Add the same number of `0`s to the **right** side
##### 5. Convert the result back to int

##### Formula : `num * 2 ** number_of_bits`

```python
print('bin(8)  =', bin(8))   # 0b1000     → 32-bit: 00000000000000000000000000001000
print('bin(56) =', bin(56))  # 0b111000   → 32-bit: 00000000000000000000000000111000
print('bin(23) =', bin(23))  # 0b10111    → 32-bit: 00000000000000000000000000010111
print()

# Example 1
a = 56
b = 23
c = a << b
print('Bitwise Left Shift of 56 << 23 =', c)  # 0b11100000000000000000000000000
print()

# Example 2
d = 34
e = 12
f = d << e
print('bin(34) =', bin(34))                      # 0b100010
print('Bitwise Left Shift of 34 << 12 =', f)     # 0b100010000000000000
print()
```

```python
# Practice - Left Shift using Formula
# Formula : num * 2 ** number_of_bits

x = int(input('Enter the Binary value  = '))
print(f'Binary number = {x} ({bin(x)})')

y = int(input('Enter the number of bits = '))
print('Number of bits = {} ({})'.format(y, bin(y)))

z = x * 2 ** y
print('Bitwise Left Shift of {} << {} = {}'.format(x, y, z))
print()
```

### 6. Bitwise Right Shift - `>>`

##### It shifts the binary digits to the **right** by the specified number of bits
##### Syntax : `number >> number_of_bits`

### Steps:
##### 1. Convert the number to binary digits
##### 2. Represent the number in a 32-bit system
##### 3. Remove the specified number of bits from the **right** side
##### 4. Add the same number of `0`s to the **left** side
##### 5. Convert the result back to int

##### Formula : `num // 2 ** number_of_bits`

### Manual Calculation Example:
##### bin(122) = 1111010 → 32-bit: 00000000000000000000000001111010
##### bit = 3           → 00000000000000000000000000001111  (removed 3 bits from right, added 3 zeros on left)
##### Result = 0b00000000000000000000000000001111 = 15

```python
a = int(input('Enter the Binary value = '))
print('Binary value = {} ({})'.format(a, bin(a)))

b = int(input('Enter the Bit value = '))
print(f'Bit value = {b} ({bin(b)})')

c = a >> b
print('Bitwise Right Shift of {} >> {} = {}'.format(a, b, c))
print()
```

```python
# Practice - Right Shift using Formula
# Formula : num // 2 ** number_of_bits

a = int(input('Enter the Binary value = '))
print(f'Binary value = {a} ({bin(a)})')

b = int(input('Enter the Bit value = '))
print('Bit value = {} ({})'.format(b, bin(b)))

c = a // 2 ** b
print(f'Bitwise Right Shift of {a} >> {b} = {c}')
```
