```python
# Functions.
# - A function provides us the facility that we write the code once and use it as and when required in a program without writing the code agian.
# - A function provides us the facility of code reusability.
# - A function is a block of code that gets executed as and when it is called.

# Types - 2:
# 1. Built-in functions
# 2. User defined functions.
```

```python
# 1. Built-in functions.

# - The functions that are predefined in python are the built-in functions.
# - We can not see the code behind those functions but we can use them.

# Ex: - input(), print(), len(), type(), ord(), id(), dir(), strip(), split() etc.....

# 2. User defined functions.

# - The functions that are defined by the programmer are the user defined functions.
# - def keyword is used to define a user defined function.

# def xxx():  ----> xxx is the function name, which should be valid identifier.
```

```python
# components of a function - 2.

# 1. Function definition.
# 2. Function call

# 1. function definition.
# - This is defining the function name and logic (code).
# - A function is defined with def keyword.

def hi():
    print('Python')  # whole unit is function definition and hi is the function name.

hi() # when we run the defined function name, then it will run the full code.
print()

# 2. Function call
# - A function definition doesn't get executed on its own.
# - A function call acts as trigger to function definition.
# - A function definition only gets executed as and when it is called.
# - A function should be defined first before we call it.

hi()
print()
add()  # we calling the function which we defined already..
```

```python
# Parameterised and Non-Parameterised functions.

# 1. Parameterised functions.
# - The function that takes a parameter(s) is a parameterised function.

# Paramter
# - The variables that we define inside the () of the function definition are known as parameters.

def pra(a,b):
    pass

# Argument
# - The values or variable that we define inside the () of the function call are known as arguments.

# def pra(a,b): // a and b are parameters.

# pra(5,6) where 5 and 6 are the value of a and b.

# parameters and arguments are used interchangably often.

# 2. Non - Parameterised 
# - A function that does not takes any parameter is a non-parameterised function.

def tes():  # no parameter or argument inside the function.
    pass
```

```python
# Parameterised Vs Non-Parameterised functions.

# create a function greet that prints have a good day! on the screen.

def greet():  # Non-Parameterised
    print('Have a good day!')

greet()
print()

a = int(input('Enter a : '))
print('Enter a: ',a)
b = int(input('Enter b : '))
print('Enter b: ',b)

def add(a,b):  # Parameterised.
    c = a + b
    print(f'a({a}) + b({b}) = {c}')

add(a,b)
print()
```

```python
# Global and Local Varaiables.

# Global Variable.
# - The variables that we define outside a function definition are the global variables.
# - Global variables has the scope outside the function as well as inside the function.

# Local Variables.
# - The variables that are defined inside a function definition are the local variables.
# - The local variables have the scope only within the function which they have been defined inside.
# - The local variables have no scope outside the function.

a = 10 # global variable. - we can access and use it further, even we can access or use it locally.

def vvv():
    x = 20 # local variable. - we have to use inside the function definition only.

```

```python
# print vs return statement.

# - print() : only prints the value where it does not return anything.
# - if any function does not return anything then it returns None by default.

# - return statement returns the value from the function back to the function call.

a = int(input('Enter a : '))
print('Enter a : ',a)
b = int(input('Enter b : '))
print('Enter b : ',b)

def add (a,b):
    c = a + b
    # print(c)
    return c

x = add(a,b)
print()

lst = []
lst.append(x)
print(lst)
```
