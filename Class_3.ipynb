```python
## Translator
#### Small software which converts the HLL to LLL (vice versa)

## Types
#### 1. Compiler.
#### 2. Interpreter.

## 1. Compiler 
#### Whole source code is converted to machine code at once, and then executed.
#### Examples: C, C++, Rust.

## 2. Interpreter
#### Source code is converted to machine code line by line, and executed immediately. 
#### Examples: Python, JavaScript.

### Compiled and Interpreted languages
#### Some languages can be both compiled and interpreted, depending on the implementation.
#### Examples: Java (compiled to bytecode, then interpreted by JVM), Python (can be compiled to bytecode, but often interpreted).
```

```python
#### How python code works:
#### First, the compiler reads all the line of source code. 
#### Then, it converts into intermediate code (bytecode). which we can't read or see.
#### Finally, the interpreter reads the bytecode and executes it line by line and converts it into machine code for execution.
#### Source Code → Bytecode → Machine Code → Output ✅

#### Python is an interpreter or compiler language?
#### Python is both a compiled and interpreted language. The source code is first compiled into bytecode (intermediate code), which is then interpreted by the Python Virtual Machine (PVM) and converted into machine code for execution.

#### Errors:
#### Syntax errors stop execution before anything runs.
#### Runtime errors allow earlier lines to run, then halt at the error.

## Types of errors:

#### Compile errors : error occurs during compilation.
#### Runtime errors : error occurs during execution.  

#### 1. Compile time errors: syntax errors and Indentation errors.
#### 2. Runtime errors: type errors, name errors, index errors, etc..
```
