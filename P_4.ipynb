```python
# Q1. Check if a number is Positive, Negative or Zero
# Enter number: -5
# -5 is Negative

x = int(input('Enter a number : '))
print('Enter a number : ',x)
print()

if x > 0:
    print(f'{x} is Positive')
elif x < 0:
    print(f'{x} is Negative')
else:
    print(f'{x} is zero')
```

```python
# Q2. Check if a person can Vote
# Enter age: 17
# Sorry! You cannot vote
# Enter age: 18
# Congratulations! You can vote

v = int(input('Enter your age : '))
print('Enter your Age : ', v)
print()

if v >= 18:
    print('🙌🤩 Congratulations! You can vote 🤩🙌')
else:
    print('😒😣 Sorry! You cannot vote 😣😒')
```

```python
# Q3. Find Grade based on marks
# 90-100 → A
# 80-89  → B
# 70-79  → C
# 60-69  → D
# Below 60 → Fail

m = int(input('Enter your marks : '))
print('Enter your Marks : ',m)
print()

if m >= 90 and m <=100:
    print('Your Grade is : A')
elif m >= 80 and m <90:
    print('Your Grade is : B')
elif m >= 70 and m <80:
    print('Your Grade is : c')
elif m >= 60 and m <70:
    print('Your Grade is : D')
elif m >= 0 and m <60:
     print('Sorry You are Fail')
else:
    print('Enter a valid Marks')
```

```python
# Q4. Check if a number is Even or Odd
# Enter number: 7
# 7 is Odd

n = int(input('Enter n : '))
print('Enter number : ',n)
print()

if n % 2 == 0:
    print(f'{n} is Even number')
else:
    print(f'{n} is odd number')
```

```python
# Q5. Simple Calculator (take two numbers and operator)
# Enter a: 10
# Enter b: 5
# Enter operator (+,-,*,/): +
# 10 + 5 = 15

a = int(input('Enter a : '))
print('Enter number : ',a)
b = int(input('Enter b : '))
print('Enter number : ',b)
print()

o = input('Enter operator: ')
print('Enter Operator : ',o)
print()

if o == '+':
    c = a + b
    print(f'{a} {o} {b} = {c}')
elif o == '-':
    c = a - b
    print(f'{a} {o} {b} = {c}')
elif o == '*':
    c = a * b
    print(f'{a} {o} {b} = {c}')
elif o == '/':
    c = a / b
    print(f'{a} {o} {b} = {c}')
elif o == '//':
    c = a // b
    print(f'{a} {o} {b} = {c}')
elif o == '%':
    c = a % b
    print(f'{a} {o} {b} = {c}')
else:
    print('Enter a Valid Operator')
```
