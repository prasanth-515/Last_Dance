### Nested Loop

```python
# nested Loop
# when there  is a loop within another is called as nested loop.

# syntax:
# for loop: #outer loop
#   for loop: #inner loop
#       code
#       code

# - for each iteration of the outer loop, the inner loop completes all its iterations.

for a in range(2):           # in total 6 times print 'Hai'
    for b in range(3):       # where the inner loop runs 3 times and whole 3 times of inner loop runs twice.
        print('Hai',end='_')    
```

```python
names = ['Prasanth', 'Praven', 'Sam', 'Ravi', 'Kumar', 'Priya', 'Ram', 'Divya', 'Sanjay','Vijay']

vowels = ''
for a in names:
    print(a)
    for b in a:
        if b in 'aeiouAEIOU':
            vowels=vowels + b
print()
print(vowels)
```

```python
lt = [(1,4,6,),(3,9,0,8),(2,7,8,1),(2,5,6,7)]

Total = 0
tt = []

for a in lt:
    print(a)
    for b in a:
        Total = Total+b
    tt.append(Total)
print(tt)
print(Total)
```

### While Loop

```python
# for loop is used with iterables only.
# for loop always has the definite number of iterations.

# while loop is used without iterable.
# while loop can have infinite iterations as well. 
# whenever we don't have the number of iterations, in such a case we use while loop.

# Components of while loop.
# 1. Counter initialisation.
# 2. Condition
# 3. Increment or Decrement
```

```python
# print the numbers from 1 to 5 using while loop.

n = 1               # counter intialisation
while n<=10:       # condition
    print(n)
    n=n+1           # increment or decrement
```

```python
# print even numbers from 2 to 20

n = 1

while n<=20:
    if n%2==0:
        print(n,end='_')
    n=n+1
```

```python
a = 'Python'
b = 0
while b<len(a):
    print(a[b],[b])
    b=b+1
print()

c = len(a)-1
while c >=0:
    print(a[c],[c])
    c=c-1
print()

d = -1
while d >= -len(a):
    print(a[d],[d])
    d = d - 1
```

```python
a = 'Python'
b = 0
r = 1
while b < len(a):
    print(a,a[b],[b],a[b]*r)
    b = b+1
    r = r+1
print()

for x in range(len(a)):
    print(a[x],x)
print()

for y in range(-1,-(len(a)+1),-1):
    print(a[y],y)

```

```python
# Transfer statements

# Break & Continue

# 1. Break.
# - break is a statement which terminates a loop.
# - as soon as break statement gets executed, the control comes out of the loop.

```

```python
for i in range(2):
    print('Hi')
    for i in range(3):
        print('hello')
        break
    break
```

```python
# Continue

# it skips the current iteration.
# it doesn't terminate the loop just skips the iteration.
# As soon as continue gets executed, the control goes to the next iteration.

a = [1001,1002,1003,1004,None,1005,1006]

for i in a:
    if i==None:
        continue
    else:
        print(i)
```

```python
# pass

# pass is an empty statement
# whenever we don't have any code or statement to keep inside an indented block then we use 'pass' to avoid error.

if 2==2:
    pass
```

```python
# Transfer Statements.

# 1. Break
# 2. continue

# 1. Break
# - It terminates a loop.
# - As soon as the break gets executed, the control comes out of the loop.

a = {101,102,103,104,None,105,106}
print(len(a))
a = [101,102,103,104,None,105,106]
n = 0
while n <= len(a):
    if a[n] == None:
        break
    else:
        print(a[n])
    n = n + 1
```

```python
# 2. continue

# - It skips the current iteration.
# - It does not terminate the loop just skips the iteration. 
# - As soon as the continue executes, it just skips the current iteration.

a = (121,122,123,124,'AAA',126,127)
n = 0
while n < len(a):
    if a[n] == 'AAA':
        n = n+1
        continue    
    else:
        print(a[n])
    n = n + 1 
```
