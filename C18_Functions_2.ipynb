```python
# Types of Arguments.

# 1. Requires Positional Arguments.
# 2. Default Arguments.
# 3. Keyworded Arguments.
# 4. Variable Length Arguments.
# 5. Keyworded Variable Length Arguments.
```

```python
# 1. Requires Positional Arguments.

# - The number of parameters and the number of arguments should be equal.
# - The values are passed from the arguments to parameters as per the order they are positioned.
# - Order matters, name does not matter.

def course (a,b,c):
    print(a)
    print(b)
    print(c)

course('DS','DA','BA')
print()

def course (a,b,c):
    print(a)
    print(b)
    print(c)
a = 'BA'
b = 'DA'
c = 'DS'

course(c,b,a)

x = 'Tamil'
y = 'English'
z = 'Maths'

def course (a,b,c):
    print(a)
    print(b)
    print(c)

course(x,y,z)
```

```python
# 2. Default Arguments.

# - When we assign a value to a parameter inside the function definition then we it becomes a default argument.
# - If don't pass any value from the function call for the default argument then it takes default value.
# - If we pass some value from the function call for the default argument then it takes the value passed to it.
# - Default arguments should be placed after the positional arguments.

def aur(f,g,h=10): # a and b are the positional arguments, c is default argument.
    print(f,g,h)  # default argument should always come at the end.

aur(11,12)
aur(8,9,13)
```

```python
# 3. Keyworded Arguments.
# - When we assign some value to a variable inside the function call then it becomes a keyworded argument.
# - The values are passed as per the name not as per the order.
# - Keyworded arguments should be placed after the positional arguments.

# def zve(i,j,k=30): # k is the keyworded argument.

def zve(i,j,k):
    print(i)
    print(j)
    print(k)

zve(1,2,k=3)
print()

zve(33,k=2,j=100) # the values are ordered based on the variable which is assigned.
print()

# zve(23,78,j=190) // which is error bcs j have 2 parameters where one positional and default argument.

# zve(12,j=123,3) // which is syntax error where the default argument should comes at the end.
```

```python
# 4. Variable Length Arguments.

# - When we don't know how many arguments are there then we use variable length arguments.
# - It internally creates a tuple and stores all the arguments passed.

# def uuu(*l): # l is a variable length argument.
#     pass

def uuu(*l):
    print(l,type(l))

uuu() # we can pass how many ever argument we want.
uuu(1,2,3,4,4)
# Create a function add_nums that can add n numbers given by the user.

# m = int(input('Enter m : '))
# print('Enter m: ',m)
def ooo(*m):
    l = 0
    for i in m:
        l = l + i
    return l

ooo(2,3,6,5,43,1)
```

```python
# 5. Keyworded Variable Length Arguments.

# - When we don't know how many keyworded arguments are there then we use Keyworded Variable Length Arguments.
# - It internally stores the values in a dictionary.

# def eiow(**v): v is a keyworded variable length argument

def eiow(**v):
    print(v,type(v))

eiow( name = "Prasanth", age = 34, course = "DS", phone = None , Roll_no = 123) 

# which will automatically converts into dict.
# where we no need to make a quotes to the keys. everything should be in the form of key =  value.
```

```python
# Anonymous function.

# - A function without any name is an anonymous function.
# - Lambda keyword is used to define a an anonymous function.
# - It is a one liner code to create a function.

# (lambda argument : expression) (arguments) --> syntax
 # create a lambda() that adds two numbers.

o = (lambda x,y : x+y) (3,5)
print(o)
print()
```

```python
# Function aliasing
# - When we assign a function to a variable then we call it function aliasing.
 
add = lambda e,v : e+v  # add is the function aliasing.
print(f'{add(3,7)}')
print()

arithmetic = lambda x,y : [x+y,x-y,x*y,x/y]
print(arithmetic(2,6))
```

```python
# Conditions inside the lambda()

# - The condition are passed using the ternary operator.

# Create a lambda() that takes a number and return 'even' or 'odd'

odd_even = lambda x : 'even' if x % 2 == 0 else 'Odd'

l = int(input('Enter l : '))
print('Enter l : ',l)

odd_even(l)
```

```python
# Maximum of 2 numbers.

ma =  lambda x,y : f'x({x}) is greater' if x > y else f'y({y}) is greater'

x = int(input('Enter x : '))
print('Enter x : ',x)
y = int(input('Enter y : '))
print('Enter y : ',y)
print()

print(ma(x,y)) 
```

```python
# let even be squared and odd be cubed

sc = lambda q : q ** 2 if q % 2 == 0 else q ** 3

q = int(input('Enter q : '))
print('Enter q : ',q)
print()

print(sc(q))
```

```python
# filter()

# - It takes a function and an iterable as arguments and returns a filter object.
# - It filter the values from an iterable.
# - It calls the function on each element of the iterable.
# - For whatever values the function returns True, it stores those values inside the filter object.
# - We can typecast the filter object to list or tuple to get the output.

# filter (function, iterable)
```

```python
lst = [33,56,23,63,44,12,75,28,49,28]

res = []
for i in lst:
    if i % 2 == 0:
        res.append(i)

print(res)
print()

ava = filter (lambda x : x % 2 == 0,lst)
print(ava) # we need to typecast to get the result
print()
re = list(ava)
print(re)
print()
```

```python
def qwe(x):
    if x % 2 == 0:
        return True
    else:
        return False

lst = [33,56,23,63,44,12,75,28,49,28,33,56,23,63,44,12,75,28,49,28]

l = []
for i in lst:
    if qwe(i):
        # print(i)
        l.append(i)
print(l)

ve = list( filter (qwe,lst) )
print(ve)
```

```python
# map()

# - It takes a function and an iterable or more iterables as arguments and returns a map object.
# - It is used to perform element by element operation on the iterable.
# - We can typecast map object to list() or tuple()

wet = [2,5,7,2,1,4,5,9,0,3,1]

qop = []
for i in wet:
    qop.append(i**2)
print(qop)
print()

rt = list(map(lambda x : x**2 , wet ))
print(rt)
```
