### FILE HANDLING

```python
# File
# - A file is a disk space that we use to store some data.

# File Handling.
# - Performing file related operations using Python is known as file handling.

# Operations on a file.
# - Generally we perform CRUD operations on files.
# - CRUD : Create Read Update Delete.
```

```python
# Types of files in python.

# 1. Text files.
# 2. Binary files.
```

```python
# 1. Text files.
# - The files that store the data in text format are the text files.
# Ex: .txt, .py, .pdf, .csv etc....

# 2. Binary files.
# - IF it is not a text file then it is a binary file in python.
# - These files store the data in bytes format.
# Ex: .bin, .dat, Image, Audio, Video etc...
```

```python
# Operations on a file in python.
# - Opening a file.
# - reading from a file.
# - writing into a file.
# - Closing a file.
# - Deleting a file.
```

```python
# Opening a file.
# - In python a file must be opened first.
# - open() is used to open a file
# - syntax : open('file_name.extension','mod')

# Closing a file.
# - A file must be closed after the operations are performed.
# - close() is used to close a file.
```

```python
# Functions to perform read and write operations.
# - read()
# - It reads a file from the cursor's position to the end of the file and returns a string.
# - After read() reads the file then the cursor's position is at the end of the file.

# read(int)
# - It reads the specified number of characters from the cursor's position.

# readline()
# - It reads a line at a time.
# - It reads from the cursor's position to the end of the line and returns a string.

# readlines()
# - It reads the file line by line and returns a list.

# write()
# - It takes a string as argument and writes it into a file.

# writelines()
# - It takes a list of strings and writes it into a file.

# tell()
# - It returns the cursor's current position.

# seek()
# - It shifts the cursor at the specified index.

# Note:
# - All the above functions are invoked using the file object.
```

```python
# Modes on text files.

# 1. Read mode -'r'

# - If the file already exists then it opens the file and places the cursor at the begining of the file.
# - If the file does not exist then it raises FileNotFoundError.
# - We are allowed to read from a file but we are not allowed to write into it.
# - The default mode is read mode.
```

```python
# Read mode.

f = open('C21.1_Intro.txt','r')
print(f.tell())
print(f.read())
print(f.tell())
print()

fa = open('C21.1_Intro.txt')
print(fa.tell())
print(fa.read())
print(fa.tell())
print()

f.seek(7)
print(f.tell())
print(f.read())
print(f.tell())
```

```python
# 2. Write mode - 'w'

# - If a file doesn't exist then it creates the file with the given name and the extension.
# - If a file already exist then it opens as a new file and all the previous content is erased.
# - We are allowed to write into a file but we are not allowed to read from a file.

a = open('C21.2_Cartoon.txt','w')
a.write('I love cartoons.\nHow about you ?')
a.close()
a = open('C21.2_Cartoon.txt','r')
print(a.read())
print(a.tell())

f = open('C21.1_Intro.txt','w')  #if we use write mode on already exist file, then it will erase everything.
print(f.write('Hello friends'))
print(f.seek(0))

f = open('C21.1_Intro.txt','r')
print(f.tell())
print(f.read())
print(f.tell())
f.close()

```

```python
# 3. append mode - 'a'

# - If a file already exist then it opens a file in append mode and places the cursor at the end of the file. 
# - If a file does not exist, it creates the file.
# - We are allowed to write into the file but we are not allowed to read from a file. 
# - Note: seek() is not applicable while writing.
# - It means it always appends the text at the end of the file irrespective of the cursor's position.

fa= open('C21.2_Cartoon.txt','a')
fa.write('\nWhat is your favorite cartoon ?')
fa.close()

fa = open('C21.2_Cartoon.txt','a')
fa.seek(10)
fa.write('\nMy favorite cartoon is Ben 10')
fa.close()

fa = open('C21.2_Cartoon.txt','r')
print(fa.read())
```

```python
# 4. read and write mode - 'r+'

# - If a file already exist then it opens the file and places the cursor at the begining of the file. 
# - If a file does not exist then it raises FileNotFoundError.
# - We are allowed to read from the file as well as we are allowed to write into the file.
# - While writing, it overwrites any data already exist from the writing position. 

x = open('C21.2_Cartoon.txt','r+')
# print(x.tell())
# print(x.read())
# print(x.tell())
# x.write('\nI also like Dragon Booster')
# x.seek(0)
# print('=================================')
print(x.read())
print(x.tell())
x = open('C21.2_Cartoon.txt','r+')
print(x.tell())
x.write('HIIIIIIIIIIII')  # which will replace.
x.seek(0)
print(x.read())

```

```python
# 5. write and read mode - 'w+'

# - If the file does not exist then it creates the file with the given name and extension.
# - If a file already exist then it opens the file and removes the previous content.
# - We are allowed to write into the file as well as we are allowed to read from the file.

xx = open('C21.3_Practice.txt','w+')
xx.close()

xx = open('C21.3_Practice.txt','w+')
xx.close() # which will edit and clear the file, if we open the exist file and which have some data in it.

xy = open('C21.3_Practice.txt','w+')
xy.write('This is the latest file.')
xy.write('\nWhich i am currently working.')
xy.seek(0)
print(xy.read())
```
