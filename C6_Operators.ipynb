```python
### 1. Arithmetic Opertors 
#### to perform arithmetic operations

# + - addition
# - - subtraction
# * - multiplication
# / - float division 
# // - floor division (int division)
# % - modulo operator (to get the remainder value)
# ** - exponent (a**b - a raise to the power b)

a = int (input('a'))
print(f'a = {a}')

b = float(input('b'))
print('b = {}'.format(b))
print()

x = a + b
print(f'The addition of a and b is {int(x)}')
print()

y = a - b
print('a - b =', int(y))
print()

z = a * b
print('a * b =', int(z))
print()

l = a / b
print('a / b = ', l)
print()

m = a // b
print(' a // b = ', m)
print()

n = a % b
print('a % b = ', n)
print()

o = a ** b
print('a ** b', o)
print()

```

```python
#### 2. Logical Operators
#### These operators act upon the boolean values and return the boolean values after the logical operation

### and
### or
### not

#### and -  read from left to right
#### it looks for false. if we get one false then it will give false
#### if there is no false then it will give true

#### Truth Table (TT, TF, FT, FF)
print('T and T is', True and True)
print('T and F is',True and False)
print('F and T is',False and True) ## while reading from left to right, it will stop at first false itself, since it's looking only for false value
print('F and F is',False and False)

a = bool()       # empty () is false
b = bool(20)
c = a and b
print(c)
print()

#### or - oppsite and operator: it looks for true
#### if we get one true then it will give true
#### if there is no true then it will give false

#### Truth Table (TT, TF, FT, FF)
print('T or T is', True or True)
print('T or F is',True or False)
print('F or T is',False or True) ## while reading from left to right, it will stop at first true itself, since it's looking only for false value
print('F or F is',False or False)

d = bool()
e = bool(1)
f = d or e
print(f)
print()

#### not - returns the opposite.

#### Truth Table (TT, TF, FT, FF)
print('T not is', not True)
print('F not is', not False)
print()

g = bool()
h = not g
print (h)
```

```python
#### 3. Comparision Operators.

#### these operators compare the values and return the boolean values. 
#### comparision can be made between the comparitable values. 
# If comparitable is not there then it raises Type error

# < less than
# > greater than
# <= less than or equal to
# >= greater than or equal to 

# int/float <-> int/float
# string <-> string 
# int <-> string // error 

a = 34
b = 24.98
c = a < b
print(f'{a} < {b} is {c}')
x= not c
print(x)
print()

d = 87
e = 55
f = d >= e
print('{} >= {} is {}'.format(d,e,f))
y = not f
print(y)
print()


#### in case of strings -  ASCII values are compared (American Standard Code for Information Interchange.)
#### to know the ASCII values -  ord('character') //syntax
#### it just judge by comparison with first letter, if it is equal then it will goes with 2nd letter.

a = 'A'
b = ord(a)
print(f'ASCII value of {a} is {b}')

print("ASCII value of x is", ord('x'))
print()

m = "A"
n = "x" 
o = m >= n
print(o)
```

```python
#### 4. Equality Operators

# These operators check the type compatibility and values and return the boolean value
# == equal to equal to 
# != not equal to

# it checks the type compatibility first
# if the types are compatible to be compared then it checks the value 
# if the values are equal then it returns True or False
# if the types are not compatible then also it return false.

a = 34
b = 67
c = a == b
print(f'a({a}) == b({b}) is {c}')
print()

d = input('d')
print('Enter d =', d)
e = input('e')
print('Enter e =', e)
f = d==e
g = d!=e
print()

print(f'ASCII value of d ({d})')
for char in d:
    print('char({}) = {}'.format(char, ord(char)))
print()
print(f'ASCII value of e ({e})')
for char in e:
    print('char({}) = {}'.format(char, ord(char)))

print()
print('d({}) == e({}) is {}'.format(d,e,f))
print()
print(f'd({d}) != e({e}) is {g}')
```

```python
### 6. Membership Operator

# These operators check whether some value is the part of a container type or not and return a bool value.
# in 
# not in

# in
# if some value is the part of a container type then it return true else false
# container types - str, list, tuple, dict, set, frozenset, range

a = 'Python '
print('P' in a)
print('sa' in a)
print('on' in a)
print('Pyh' in a) # should be exactly as per as given
print(' ' in a)
print()

b = [ 13, 'Python', (34, 4.3, 'Hello'), {3:35, 'Age' : 34, 'Name' : 'Prasanth'}, True, False]
print(0 in b)
print()

# not in
# if a value is not the part of the container type then it returns True else False.

a = 'Python '
print('P' not in a)
print('sa' not in a)
print('on' not in a)
print('Pyh'not in a) # should be exactly as per as given
print(' ' not in a)
print()

b = [ 13, 'Python', (34, 4.3, 'Hello'), {3:35, 'Age' : 34, 'Name' : 'Prasanth'}, True, False]
print(0 not in b)
print()
```

```python
### 7. Identity Operators.

# These operators compare the memory addresses fo two values and return a bool value. 
# id(value/variable) -> It returns the memory address

# is 
# is not

# is
# if two values have the same memory address then it returns True else false.

a = 10
b = 10
print(a is b)
print(id(a), id (b))
print()

a = 2.4
b = 2.4
print(a is b)
print(id(a), id(b))
print()

# ============================================
# Quick Comparison Table: Identity in Python
# ============================================

# Type            | Example                        | is result | Reason
# ----------------|--------------------------------|-----------|----------------------------
# Small int       | 10 is 10                       | True      | Cached (-5 to 256)
# Large int       | 1000 is 1000                   | False     | New object
# Float           | 2.4 is 2.4                     | False     | Not cached
# Short string    | "hi" is "hi"                   | True      | Interned
# Long string     | "hello world" is "hello world" | False     | Not always interned
# Tuple (simple)  | (1,2,3) is (1,2,3)             | True*     | Sometimes cached (depends)
# List            | [1,2] is [1,2]                 | False     | Mutable, new object
# Dict            | {'a':1} is {'a':1}             | False     | Mutable, new object
# Set             | {1,2} is {1,2}                 | False     | Mutable, new object
# User object     | MyClass() is MyClass()         | False     | Each call creates new object

# ============================================
# Notes:
# - 'is' checks identity (same memory address)
# - '==' checks value equality (same content)
# - Immutable + small/simple values → often cached
# - Mutable or large values → new memory every time
# ============================================

x = (2,3,4,6)
y = (2,3,5,8)
print(x is not y)
print()

m = (1,2,3)
n = (1,2,3)
print(m is not n)
print()

m = [1,2,3]
n = [1,2,3]
print(m is not n)
```

```python
### 8. Bitwise Operator

# These are operators are used to perform bit level operations.
# A binary digit is known as bit.

# steps
# 1. Convert a number into binary digit known as bit
# 2. Perform the bitwise operation bit by bit.
# 3. Conver the result back to number.

# Note: Bitwise operators are only applicable on int values
# As only int values can be converted to binary digits

# & bitwise and 
# | bitwise or
# ~ bitwise not
# ^ bitwise xor
# << bitwise left shift
# >> bitwise right shift

### bitwise and - &

# and -> T and T is T else False
# & -> 1 and 1 is 1 else 0

a = 78
b = 235
c = a & b
print(c, bin(c))

print(bin(a))  # 01001110
print(bin(b))  # 11101011       # 01001010  ## adding 0b before binary number will give the integer of the respective binary number

0b01001010
print()

### bitwise or - |

# or -> f or f is f else True
# | -> 0 | 0 is 0 else 1

a = 182
b = 392                             # 010110110
print(bin(a), bin(b)) # 0b10110110 #0b110001000 = 110111110

c = a | b 
print(c)
0b110111110
print()

### bitwise not - ~
# ~num = -(num+1) //formual
a = 5
b = ~a
print('~ of 5 is', b)


### bitwise xor - ^ (caret sign)

# The same digits return 0, opposite return 1
a = 182
b = 392                             # 010110110
print(bin(a), bin(b)) # 0b10110110 #0b110001000 = 100111110

c = a ^ b 
print(c)
print()
0b100111110
print()

### bitwise leftshift - <<

# it shift the binary digits to the left by the specified number of digits.
# number << the number of bits

# steps
# 1. convert the number to binary digits 
# 2. represent the number in a 32 bit system representation.
# 3. remove the specified number of bits from the left side and add the same number of 0 to the right side.
# 4. convert the result to int.


print('bin(8)',bin(8)) #'0b1000' - stored in 32 bit system, which is stored as 000000000000000000000000000000001000 (32 - 0's with bin vlaue)
print('bin(56)',bin (56)) #00000000000000000000000000000000111000
print('bin(23)',bin(23))  
print()

a = 56
b = 23
c = a << b      #11100000000000000000000000000
print('bitwise leftshift of 56<<23', c)
0b11100000000000000000000000000

d = 34
e = 12
f = d<<e
print('bin(34)',bin(34)) # 00000000000000000000000000000000100010
print('Bitwise left shift of 34<<12', (f)) # 00000000000000000000100010000000000000
0b00000000000000000000100010000000000000
print()

# formula - num*2**the_number-of-bits


# bitwise right shift - >>

# steps
# 1. convert the number to binary digits 
# 2. represent the number in a 32 bit system representation.
# 3. remove the specified number of bits from the right side and add the same number of 0 to the left side.
# 4. convert the result to int.

a = int(input("Enter the Binary value = "))
print('Enter the Binary value = {} ({})'.format(a, bin(a)))
b = int(input('Enter the Bit value = '))
print(f'Enter the Bit value = {b} ({bin(b)})')

c = a>>b
print('Bitwise Rightshift of binary {} and bit {} is {} '.format (a,b,c))
print()

# manual calculation 
# bin of 122 1111010 -> 00000000000000000000000001111010  (32bit)
# bit is 3           -> 00000000000000000000000000001111 // removing 3 bit from right and adding 3 000 on left.
0b00000000000000000000000000001111

# formula - num//2**number-of-bit
```

```python
# Practice - Left Shift using Formula
# Formula : num * 2 ** number_of_bits

x = int(input('Enter the Binary value  = '))
print(f'Binary number = {x} ({bin(x)})')

y = int(input('Enter the number of bits = '))
print('Number of bits = {} ({})'.format(y, bin(y)))

z = x * 2 ** y
print('Bitwise Left Shift of {} << {} = {}'.format(x, y, z))
print()

# Practice - Right Shift using Formula
# Formula : num // 2 ** number_of_bits

a = int(input('Enter the Binary value = '))
print(f'Binary value = {a} ({bin(a)})')

b = int(input('Enter the Bit value = '))
print('Bit value = {} ({})'.format(b, bin(b)))

c = a // 2 ** b
print(f'Bitwise Right Shift of {a} >> {b} = {c}')
```

```python
### 9. Ternary Operator

# It is a substitute to condtional statements (if...else statement)
# It is a one liner code to implement if...else statement.

# Syntax
# exp_1 if condition else exp_2
# when the if condition is true then exp_1 is the result
# when the if condition is false then exp_2 is the result.

print('a') if 3==2 else print('b')
print()

a = int(input('Enter the value = '))
print(f'Enter the value of a = {a}')

print(f"a ({a}) is Even") if a%2==0 else print('a ({}) is odd'.format(a)) 
```
